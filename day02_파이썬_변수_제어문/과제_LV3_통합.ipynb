## 1. 숫자 맞추기 업다운 게임

**배경**: 컴퓨터가 **1~100 사이의 숫자 하나**를 정합니다. 여러분은 그 수를 **최대 7번** 안에 맞혀야 합니다.

**UP / DOWN 의 의미** (헷갈리기 쉬우니 확인하세요)
- 입력한 수가 정답보다 **작으면** → `UP` : 정답은 **더 큰 수**입니다
- 입력한 수가 정답보다 **크면** → `DOWN` : 정답은 **더 작은 수**입니다

### 단계별로 만들기
아래 **답안 셀 하나**에 작성합니다. v1부터 순서대로 키워 나가고, **제출은 v3까지 반영된 완성본 하나**입니다.
(v1·v2를 따로 남길 필요는 없습니다.)

- **v1 (기본)**: 숫자를 한 번 입력받아 정답과 비교해 `UP` / `DOWN` / 정답 중 하나를 출력
- **v2 (시도 제한)**: `for attempt in range(1, 8):` 로 **최대 7번** 반복. 맞히면 `break` 로 즉시 종료, 7번을 다 쓰도록 못 맞히면 실패 메시지와 함께 **정답을 알려 준다**
- **v3 (입력 검증)**: 1~100 **범위 밖 입력**이면 경고를 출력하고 `continue` 로 다음 회차로 넘어간다

**요구사항**
- 회차는 `[1/7]` 처럼 **현재 회차 / 전체 7회** 형식으로 표시한다
- 맞혔을 때는 **몇 번 만에** 맞혔는지 함께 출력한다
- 범위 밖 입력도 **시도 횟수를 1회 사용한 것으로 친다** (`continue` 를 써도 그 회차는 지나갑니다)
- 성공했는지 여부를 **플래그 변수**(예: `success`)로 기억해, 반복이 끝난 뒤 성공/실패를 구분한다
- 아래 **[제공 코드]** 셀 외에는 `import` 를 쓰지 않는다
- 출력 문구가 예시와 글자까지 같을 필요는 없지만, **회차 표시·UP/DOWN 방향·성공 시 시도 횟수**는 반드시 드러나야 한다

**예시** (정답이 `37` 인 경우)
```
[1/7] 1~100 사이 숫자: 50
⬇️ 더 작은 수입니다 (DOWN)
[2/7] 1~100 사이 숫자: 25
⬆️ 더 큰 수입니다 (UP)
[3/7] 1~100 사이 숫자: 37
정답! 3번 만에 맞혔습니다.
```

7번을 다 써도 못 맞힌 경우:
```
[7/7] 1~100 사이 숫자: 88
⬇️ 더 작은 수입니다 (DOWN)
실패! 정답은 37 였습니다.
```

## 2. 숫자 메뉴 자판기 시뮬레이터

**배경**: 번호로 조작하는 음료 자판기를 만듭니다. **잔액을 충전**하고, 음료를 **구매**하고,
종료 시 **거스름돈**을 돌려줍니다. `while` 메뉴 루프로 동작합니다.

### 메뉴
```
1. 콜라   (1500원)
2. 사이다 (1200원)
3. 물     ( 800원)
9. 잔액 충전
0. 종료
```

**요구사항**
- 시작 **잔액은 0원**이다
- `while` 로 **메뉴와 현재 잔액**을 반복 표시하고 번호를 입력받는다
- `9` (충전) → 충전할 금액을 입력받아 잔액에 더한다. 단, **0원 이하를 입력하면 충전하지 않고 안내**한다
- `1`/`2`/`3` (구매) → 잔액이 **가격 이상이면** 구매하고 그만큼 차감한 뒤 **남은 잔액**을 알려 준다. **부족하면 구매하지 않고** 상품 가격과 현재 잔액을 안내한다
- `0` (종료) → **남은 잔액을 거스름돈으로 반환**한다고 출력하고 반복을 끝낸다 (`break`)
- **그 밖의 번호**(예: `5`, `7`)를 입력하면 "없는 메뉴입니다" 안내 후 **다시 메뉴로 돌아간다**
- 구매·충전 후에도 **`0` 을 누를 때까지 메뉴가 계속 반복**된다
- **딕셔너리는 쓰지 않습니다.** 상품 이름·가격을 개별 변수와 `if` 로 처리하세요
- 상품 이름·가격은 위 메뉴표의 값을 그대로 씁니다 (콜라 1500 / 사이다 1200 / 물 800)

**예시** (일부)
```
1. 콜라   (1500원)
2. 사이다 (1200원)
3. 물     ( 800원)
9. 잔액 충전
0. 종료
현재 잔액: 0원
번호 선택: 1
잔액이 부족합니다. (콜라 1,500원 / 현재 0원)
...
번호 선택: 9
충전할 금액: 2000
2,000원 충전 완료
...
번호 선택: 1
콜라 나왔습니다. 남은 잔액 500원
...
번호 선택: 0
거스름돈 500원을 반환합니다. 감사합니다!
```
