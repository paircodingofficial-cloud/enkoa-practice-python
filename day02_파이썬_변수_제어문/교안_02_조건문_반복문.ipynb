# 1. 조건문 `if` — 상황에 따라 갈라지기

## 일상 비유

조건문은 **갈림길**입니다. "비가 오면 우산을 챙기고, 아니면 그냥 나간다."
컴퓨터도 `조건이 참이면 이 코드를, 아니면 저 코드를` 실행하게 만들 수 있어요.

## 문법 — 들여쓰기가 문법의 일부!

```python
if 조건1:
    조건1이 참일 때 실행    # 반드시 들여쓰기(보통 space 4칸)
elif 조건2:
    조건1은 거짓, 조건2가 참일 때
else:
    위 조건들이 모두 거짓일 때
```

- 조건 끝에 **콜론 `:`** 을 꼭 붙입니다.
- `if` 에 딸린 코드는 **들여쓰기**로 표시합니다. (파이썬은 들여쓰기로 "묶음"을 구분!)
- `elif`, `else` 는 **없어도 됩니다.**

````python
# 가장 단순한 if - else
age = 20

if age >= 19:
    print("성인입니다 ")
else:
    print("미성년자입니다")
````

### 짝수·홀수 판별 — `%` 와 조합

앞서 배운 나머지 `%` 가 여기서 빛을 발합니다. `number % 2 == 0` 이면 짝수!

````python
number = 7

if number % 2 == 0:
    print(f"{number}은(는) 짝수")
else:
    print(f"{number}은(는) 홀수")
````

### `elif` 사다리 — 성적 등급 매기기

조건이 여러 개일 땐 `elif` 를 이어 붙입니다. **위에서부터 차례로** 검사하고,
**처음 참이 되는 곳** 하나만 실행한 뒤 빠져나옵니다.

````python
score = 83

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"{score}점 → {grade} 등급")
````

> ⚠️ **순서가 중요!** `score >= 60` 을 맨 위에 두면 90점도 D가 되어버립니다.
> `elif` 사다리는 보통 **좁은 조건(큰 값)부터** 위에 둡니다.

### 한 줄로 쓰는 조건 — 3항 연산자

`if / else` 로 **값 하나를 고르기만** 할 때는 한 줄로 줄여 쓸 수 있습니다.
정식 이름은 **조건 표현식**, 흔히 **3항 연산자**라고 부릅니다.

```
(참일 때 값) if (조건) else (거짓일 때 값)
```

영어 문장처럼 읽으면 순서가 자연스럽습니다 — "성인으로 한다, 19살 이상이면, 아니면 청소년으로".

````python
age = 20

# 지금까지 쓰던 방식
if age >= 19:
    ticket = "성인"
else:
    ticket = "청소년"
print(ticket)

# 같은 일을 한 줄로 — 3항 연산자
ticket = "성인" if age >= 19 else "청소년"
print(ticket)
````

````python
# f-string 안에서 바로 쓰기도 합니다
number = 7
print(f"{number}은(는) {'짝수' if number % 2 == 0 else '홀수'}입니다")
````

> **언제 쓰나요?** 값 하나를 고를 때만 쓰세요.
> 조건이 복잡하거나 실행할 일이 여러 줄이면 **원래의 `if / else` 가 훨씬 읽기 좋습니다.**

### 조건 결합과 중첩

`and` / `or` 로 조건을 엮거나, `if` 안에 또 `if` 를 넣어(중첩) 세밀하게 판단할 수 있습니다.

````python
# and 로 결합: 놀이기구 탑승 조건 (키 130 이상 그리고 나이 10 이상)
height_cm = 140
age = 12

if height_cm >= 130 and age >= 10:
    print("탑승 가능 ✅")
else:
    print("탑승 불가 ❌")
````

````python
# 중첩 if: 회원인지 먼저 보고, 회원이면 등급을 따짐
is_member = True
points = 1500

if is_member:
    if points >= 1000:
        print("VIP 쿠폰 지급")
    else:
        print("일반 쿠폰 지급")
else:
    print("먼저 회원 가입이 필요합니다")
````

### `in` — "안에 들어 있나?" 묻기

`in` 은 **왼쪽 값이 오른쪽 안에 들어 있는지**를 `True` / `False` 로 알려 줍니다.
문자열에 쓰면 **그 글자나 단어가 포함되어 있는지** 확인합니다. 반대는 `not in` 입니다.

조건문과 묶어서 아주 자주 쓰는 형태예요.

````python
email = "hong@enkoa.com"

# 문자열 안에 그 글자가 있는지 → True / False
print("@" in email)      # True
print("!" in email)      # False
````

````python
# 조건문과 함께
if "@" in email:
    print("이메일 형식으로 보입니다")
else:
    print("@ 가 없어 이메일이 아닙니다")
````

````python
# not in : 들어 있지 '않은지'
password = "abc123"
if "!" not in password:
    print("특수문자(!)를 넣으면 더 안전합니다")
````

````python
# 한 글자가 아니라 단어로도 확인할 수 있습니다
message = "오늘 회의는 3시입니다"
if "회의" in message:
    print("일정 관련 메시지네요")
````

# 2. 반복문 `while` — 조건이 참인 동안 계속

## 언제 쓰나요?
"**몇 번 할지 모르지만**, 어떤 조건이 만족될 때까지 계속" 반복할 때. (예: 로그인 성공할 때까지)

## 문법
```python
while 조건:
    반복할 코드      # 조건이 참인 동안 계속 실행
    ...             # 언젠가 조건이 거짓이 되도록 '값을 바꿔줘야' 함!
```

````python
# 1부터 5까지 세기
i = 1
while i <= 5:
    print(i, end=" ")
    i += 1           # ← 이 줄이 없으면 무한 반복!
print()              # 줄바꿈
````

````python
# 1부터 10까지 합 구하기 (누적 패턴)
total = 0
n = 1
while n <= 10:
    total += n        # 합계에 계속 더한다
    n += 1
print("1~10의 합:", total)   # 55
````

## 무한 루프와 `break`

조건이 **영원히 참**이면 프로그램이 멈추지 않습니다(무한 루프). 일부러 `while True:` 로 무한 반복을 만들고,
**특정 조건에서 `break` 로 탈출**하는 패턴을 아주 많이 씁니다.

> 주피터에서 무한 루프에 빠지면 위쪽 **■(정지)** 버튼으로 멈출 수 있어요.

````python
# while True + break : 시도 횟수 제한이 있는 안전한 무한 루프
chances = 3
while True:
    print(f"남은 기회: {chances}")
    chances -= 1
    if chances == 0:
        print("기회를 모두 사용했습니다.")
        break         # 여기서 while 을 완전히 빠져나감
````

## `continue` — 이번 회차만 건너뛰기

`break` 는 반복을 **완전히 끝내지만**, `continue` 는 **이번 바퀴만 건너뛰고** 다음 반복으로 갑니다.

````python
# 1~10 중 홀수만 출력 (짝수는 continue 로 건너뜀)
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue      # 짝수면 아래로 안 내려가고 다음 바퀴로
    print(i, end=" ")
print()
````

# 3. 반복문 `for` + `range()` — 정해진 횟수만큼

## 언제 쓰나요?
"**정확히 N번**" 또는 "**어떤 범위의 숫자마다**" 반복할 때. `range()` 로 숫자 범위를 만듭니다.

## `range()` 세 가지 형태

| 형태 | 뜻 | 만드는 숫자 |
|---|---|---|
| `range(끝)` | 0부터 끝-1까지 | `range(5)` → 0,1,2,3,4 |
| `range(시작, 끝)` | 시작부터 끝-1까지 | `range(1, 6)` → 1,2,3,4,5 |
| `range(시작, 끝, 간격)` | 간격만큼 건너뛰며 | `range(0, 10, 2)` → 0,2,4,6,8 |

> ⭐ **끝 숫자는 포함하지 않습니다!** (`range(1, 6)` 은 6을 포함하지 않음 → 1~5)

````python
# 형태 1: range(끝)
for i in range(5):
    print(i, end=" ")     # 0 1 2 3 4
print()
````

````python
# 형태 2: range(시작, 끝)
for i in range(1, 6):
    print(i, end=" ")     # 1 2 3 4 5
print()
````

````python
# 형태 3: range(시작, 끝, 간격)
for i in range(0, 11, 2):
    print(i, end=" ")     # 0 2 4 6 8 10
print()
````

````python
# 거꾸로도 가능 (간격을 음수로)
for i in range(10, 0, -1):
    print(i, end=" ")     # 10 9 8 ... 1
print()
````

### 구구단 한 단 출력

````python
dan = 3
for i in range(1, 10):        # 1부터 9까지
    print(f"{dan} x {i} = {dan * i}")
````

## 누적 패턴 3종 — 합 · 개수 · 최댓값

반복문의 진짜 힘은 **하나씩 훑으며 값을 쌓는** 데 있습니다.
(참고: `sum()`·`max()` 같은 편의 함수는 다음 단원에서 배웁니다. 지금은 **직접 `if` 로 추적**하는 힘을 기릅니다.)

````python
# 합: 1~100 의 합
total = 0
for i in range(1, 101):
    total += i
print("1~100 합:", total)      # 5050
````

````python
# 개수: 1~50 중 3의 배수의 개수
count = 0
for i in range(1, 51):
    if i % 3 == 0:
        count += 1
print("1~50 중 3의 배수 개수:", count)
````

````python
# 최댓값: score의 최댓값 구하기
max_value = 0
for value in range(1, 8):
    score = value * 7 % 11

    # score와 max_value 비교 
    if score > max_value: 
        max_value = score
print("가장 큰 값:", max_value)
````

### for 에서의 `break` / `continue`

````python
# break: 3의 배수를 처음 만나면 멈춤
for i in range(1, 20):
    if i % 3 == 0:
        print(f"{i} 에서 멈춤!")
        break
````

````python
# continue: 3의 배수만 건너뛰고 나머지 출력
for i in range(1, 11):
    if i % 3 == 0:
        continue
    print(i, end=" ")
print()
````

## for 에 문자열 넣기 — 글자 하나씩 순회

`for` 에 **문자열**을 넣으면, 글자를 **왼쪽부터 하나씩** 꺼내 줍니다. `range()` 없이도 반복이 되지요!

- 리스트는 다음 단원에서 배우지만, 문자열 순회는 지금 배운 `for` · `in` · `len()` 만으로 충분합니다.
- 누적 패턴(개수 세기)도 문자열에 그대로 쓸 수 있어요.

> ⚠️ 아직 **인덱싱(`글자[0]`) · 슬라이싱 · 문자열 메서드(`.upper()` 등)** 는 배우지 않았습니다.
> 여기서는 **순회**, **`in`(포함 여부)**, **`len()`(길이)**, **`+` · `*`** 만 사용합니다.

````python
# (1) 이름을 한 글자씩 출력
name = "파이썬"
for ch in name:
    print(ch)
````

````python
# (2) 문장에서 특정 글자의 개수 세기 (누적 패턴 재활용)
sentence = "banana"
count = 0
for ch in sentence:
    if ch == "a":
        count += 1
print(f"'a' 는 {count}번 나옵니다")   # 3
````

````python
# (3) in 과 결합 — 모음(a, e, i, o, u) 개수 세기
word = "watermelon"
vowel_count = 0
for ch in word:
    if ch in "aeiou":
        vowel_count += 1
print(f"모음은 {vowel_count}개, 전체 글자 수는 {len(word)}개")
````

# 4. 중첩 루프 — 반복 속의 반복

`for` 안에 또 `for` 를 넣으면, 바깥이 한 바퀴 돌 때마다 안쪽이 **전체를 다 돕니다.**
표(가로×세로)나 도형을 그릴 때 필수예요.

````python
# 구구단 전체 (2단 ~ 9단)
for dan in range(2, 10):
    for i in range(1, 10):
        print(f"{dan}x{i}={dan*i}", end="  ")
    print()          # 한 단이 끝나면 줄바꿈
````

````python
# 별로 사각형 그리기 (가로 4, 세로 3)
for row in range(3):
    for col in range(4):
        print("*", end="")
    print()
````

## 🚀 응용 클론코딩 ④ — 뱀 모양(지그재그) 숫자판

`1` 부터 `16` 까지를 **4 × 4 격자**에 채우는데, 줄마다 **방향이 번갈아** 바뀝니다.
좌우로 왔다 갔다 하는 모양이 뱀 같다고 해서 **지그재그(뱀 모양)** 라고 부릅니다.

```
 1  2  3  4     →
 8  7  6  5     ←
 9 10 11 12     →
16 15 14 13     ←
```

홀수 줄은 **왼쪽 → 오른쪽**, 짝수 줄은 **오른쪽 → 왼쪽** 으로 채워집니다.
그래서 2번째 줄의 첫 칸이 `5` 가 아니라 `8` 이에요.

### 설계
1. 한 변의 크기 `size = 4` 를 정한다
2. 줄 번호 `row` 를 `1` 부터 `size` 까지 반복한다
3. **홀수 줄**(`row % 2 == 1`)이면 칸 번호 `col` 을 `1 → size` 순서로
4. **짝수 줄**이면 `range(size, 0, -1)` 로 **거꾸로** (`range` 세 번째 인자에 음수!)
5. 각 칸에 들어갈 값은 `(row - 1) * size + col`
6. 한 줄을 문자열로 모았다가 줄 단위로 출력한다

> 표처럼 자리를 맞추려면 f-string 에 **폭**을 주세요. `f"{value:>3}"` 은 **3칸 폭, 오른쪽 정렬**입니다.

|---|---|---|
| `if / elif / else` | 조건에 따라 갈라짐, `:` 와 들여쓰기 | `if score>=90:` |
| 조건 만들기 | 비교(`==`,`>=`) + 논리(`and`,`or`,`not`) | `height_cm>=130 and age>=10` |
| 중첩 `if` | 조건 안의 조건 | 회원 → 등급 |

| `in` / `not in` | 포함되어 있는지를 `True`/`False` 로 | `if "@" in email:` |
| `while` | 조건이 참인 동안 반복 | `while i<=5:` |
| 무한 루프 | `while True:` + `break` 탈출 | 로그인 시도 제한 |
| `continue` | 이번 회차만 건너뛰기 | 짝수 건너뛰기 |
| `for` + `range` | 정해진 횟수 반복, 끝값 미포함 | `range(1,6)`→1~5 |
| `for` + 문자열 | 문자열을 `for` 에 넣어 글자 하나씩 순회 | `for ch in "abc":` |
| 누적 패턴 | 합·개수·최댓값을 `if` 로 직접 추적 | `total += i` |
| 중첩 루프 | 반복 속의 반복 | 구구단·별 도형 |
