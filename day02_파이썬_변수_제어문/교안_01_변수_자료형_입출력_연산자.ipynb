# 1. 변수 (Variable) — 값에 이름표 붙이기

## 왜 필요할까요?

계산기로 `173 + 25` 를 구했다고 합시다. 이 결과를 나중에 또 쓰려면 **어딘가에 저장**해 둬야겠죠.
매번 숫자를 다시 적는 대신, **이름을 붙여 보관**하는 것이 변수입니다.

## 일상 비유

변수는 **이름표가 붙은 상자**입니다.

- `age = 25` → "나이"라고 적힌 상자에 숫자 25를 넣는다
- `age` 라고 부르면 → 상자를 열어 안에 든 25를 꺼내 쓴다
- 상자 안 내용물은 **언제든 바꿀 수 있습니다** (재할당)

## 문법

```python
변수이름 = value
```

`=` 는 "같다"가 아니라 **"오른쪽 값을 왼쪽 이름에 넣어라"** 라는 뜻입니다. (같다는 `==`, 뒤에서 배워요!)

````python
# 변수 = 값에 이름표를 붙이는 것
age = 25
name = "김파이"
height_cm = 175.5

print(age)
print(name)
print(height_cm)
````

### 재할당과 값 맞바꾸기(동시 할당)

상자 속 값은 얼마든지 새 값으로 **덮어쓸 수 있습니다.** 또, 두 상자의 내용물을 **한 줄로 맞바꿀** 수도 있어요.

````python
# 재할당: 상자 속 값을 새 값으로 덮어쓰기
score = 90
print("처음 점수:", score)

score = 100
print("바뀐 점수:", score)
````

````python
# 동시 할당: 두 변수의 값을 한 번에 맞바꾸기
a = 1
b = 2
print("교환 전 → a =", a, ", b =", b)

a, b = b, a          # 오른쪽 값들을 왼쪽에 동시에 넣습니다
print("교환 후 → a =", a, ", b =", b)
````

## 변수 이름 짓기 규칙

| 규칙 | 예시 |
|---|---|
| ✅ 영문·숫자·밑줄(`_`)만 사용 (이 교재는 **영문 이름**만) | `total_price`, `student_count`, `count2` |
| ✅ 의미가 드러나게 (한 글자 남발 금지) | `n` 보다 `student_count` |
| ❌ **숫자로 시작 금지** | `2nd` (X) → `second` (O) |
| ❌ **하이픈(-)·띄어쓰기·특수문자 금지** | `my-name`, `my name` (X) |
| ❌ **예약어 금지** (`if`, `for`, `class`, `print` 등 파이썬이 쓰는 단어) | `class = 1` (X) |
| ✅ **관례: 스네이크 케이스**(`snake_case`) | `total_price` (O) / `totalPrice` (X) |

### 파이썬의 이름 규칙 — 스네이크 케이스(`snake_case`)

변수 이름은 **모두 소문자**로 쓰고, 단어와 단어 사이를 **밑줄(`_`)** 로 잇습니다.
밑줄로 이어진 모양이 뱀 같다고 해서 **스네이크 케이스**라고 부릅니다.
파이썬 공식 스타일 가이드(**PEP 8**)가 정한 방식이라 실무 코드도 대부분 이 규칙을 따르고,
이 교재의 모든 예제와 과제도 스네이크 케이스로 씁니다.

| | 예시 | 설명 |
|---|---|---|
| ✅ | `total_price`, `student_count`, `user_name` | 파이썬이 쓰는 방식 |
| ❌ | `totalPrice`, `studentCount` | **카멜 케이스** — 대문자가 낙타 등처럼 솟은 모양. 자바·자바스크립트에서 주로 쓰고 파이썬에서는 쓰지 않습니다 |
| ❌ | `TotalPrice` | **파스칼 케이스** — 파이썬에서 변수 이름으로는 쓰지 않습니다 |
| ❌ | `totalprice` | 단어 구분이 없어 읽기 어렵습니다 |

> **'에러'란?** 영어로 error — 프로그램이 규칙을 어겨 실행이 멈추는 상황을, 이 교재에서는 '에러'라고 부릅니다. (아래 ❌ 예시들을 실제로 실행하면 에러가 납니다.)

# 2. 자료형 (Data Type) — 값의 종류

## 왜 필요할까요?

숫자 `100` 과 글자 `"100"` 은 사람 눈엔 비슷해 보여도, 컴퓨터에겐 **완전히 다른 종류**입니다.
숫자끼리는 더할 수 있지만(`100 + 100 = 200`), 글자 `"100" + "100"` 은 `"100100"` 이 되죠.
그래서 값이 **어떤 종류인지** 아는 게 중요합니다.

## 기본 4가지 자료형

| 자료형 | 이름 | 예시 | 설명 |
|---|---|---|---|
| `int` | 정수 | `0`, `25`, `-7` | 소수점이 없는 수 |
| `float` | 실수 | `3.14`, `-0.5`, `2.0` | 소수점이 있는 수 |
| `str` | 문자열 | `"안녕"`, `'A'` | 따옴표로 감싼 글자 |
| `bool` | 불리언 | `True`, `False` | 참/거짓 (첫 글자 대문자!) |

값의 종류는 `type()` 으로 확인합니다.

````python
# type() 으로 자료형 확인하기
int_num = 10
real_num = 3.14
text = "파이썬"
boolean = True

print(type(int_num))     # <class 'int'>
print(type(real_num))     # <class 'float'>
print(type(text))   # <class 'str'>
print(type(boolean))   # <class 'bool'>
````

## 형변환 (Type Casting)

`int()`, `float()`, `str()` 로 자료형을 **서로 바꿀 수** 있습니다. 특히 `input()`은 항상 문자열을 주므로,
숫자로 계산하려면 형변환이 **꼭** 필요합니다 (뒤에서 다시 봅니다).

````python
# 문자열 "100" 과 숫자 100 은 다릅니다
num_str = "100"
number = int(num_str)          # 문자열 -> 정수

print(num_str + num_str)        # 문자열 이어붙이기 -> 100100
print(number + number)        # 숫자 덧셈      -> 200
````

````python
# 정수 <-> 실수 <-> 문자열
print(float(3))          # 3.0
print(str(42) + "점")     # "42점"  (숫자를 글자로 바꿔 이어붙임)
````

### 형변환 함정 3가지

교육에서 자주 실수하는 부분입니다. **눈으로 익혀두세요.**

1. `int("3.5")` 는 **에러!** — `"3.5"` 는 정수 모양이 아니라서 바로 정수로 못 바꿉니다.
   → `int(float("3.5"))` 처럼 **실수를 거쳐** 바꿔야 합니다.
2. `int(3.9)` 는 `4` 가 아니라 `3` — 반올림이 아니라 **소수점 아래를 버립니다**.

````python
# int("3.5") 는 에러라 실행하지 않고, 안전한 방법을 씁니다
print(int(float("3.5")))   # 3  (실수 3.5를 거쳐 정수로)
````

## 컴퓨터의 소수 이야기 — `0.1 + 0.2`

컴퓨터는 소수를 2진법으로 저장하는데, `0.1` 같은 값은 **정확히** 담지 못해 아주 작은 오차가 생깁니다.
버그가 아니라 **모든 컴퓨터의 공통 특징**이에요.

그래서 소수끼리 `==` 로 곧바로 비교하면 안 됩니다.
대신 **두 값의 차이가 충분히 작은지**를 봅니다. 이것을 **허용 오차 비교**라고 합니다.
차이를 구할 때는 `abs()`(절댓값 — 부호를 뗀 크기)를 써서, 어느 쪽이 크든 양수가 되게 만듭니다.

> 나중에 배울 `math.isclose()` 가 바로 이 허용 오차 비교를 대신해 주는 표준 도구입니다.
> (`import` 는 뒤 단원(모듈)에서 배웁니다)

````python
print(0.1 + 0.2)                   # 0.30000000000000004  (아주 작은 오차!)
print(0.1 + 0.2 == 0.3)            # False  — 오차 때문에 다르다고 나옵니다
````

````python
# 해결: 두 값의 차이가 충분히 작으면 '같다'고 본다 (허용 오차 비교)
diff = abs((0.1 + 0.2) - 0.3)      # abs() = 절댓값
print(diff)                        # 5.551115123125783e-17  (0에 아주 가까움)
print(diff < 0.000001)             # True  — 사실상 같다
````

````python
# round() 로 자릿수를 맞춰 비교할 수도 있지만, 자릿수 선택에 결과가 좌우됩니다
print(round(0.1 + 0.2, 2) == 0.3)   # True
print(round(0.1 + 0.2, 20) == 0.3)  # False <- 같은 값인데 자릿수만 바꿔도 뒤집힙니다
````

# 3. 입력과 출력 — `print()` 와 `input()`

## 3-1. `print()` 를 자유자재로

`print()` 는 여러 값을 한 번에 출력할 수 있고, `sep`(구분자)와 `end`(끝맺음)를 바꿀 수 있습니다.

| 옵션 | 뜻 | 기본값 |
|---|---|---|
| `sep` | 값 사이를 무엇으로 구분할지 | 공백 `" "` |
| `end` | 출력 끝에 무엇을 붙일지 | 줄바꿈 `"\n"` |

````python
print("사과", "바나나", "포도")                # 기본: 공백으로 구분
print("사과", "바나나", "포도", sep=", ")       # 쉼표+공백으로 구분
print("2024", "01", "15", sep="-")             # 2024-01-15

print("로딩중", end="")                         # 줄바꿈 안 함
print("...완료!")                               # 앞 줄에 이어서 출력됨
````

## 3-2. f-string — 문자열 서식의 최강자

문자열 앞에 **`f`** 를 붙이면, 중괄호 `{}` 안에 **변수나 식**을 바로 넣을 수 있습니다.
게다가 `{value:서식}` 형태로 **소수 자릿수·천 단위 쉼표·정렬·퍼센트**까지 지정할 수 있어요.

| 서식 | 뜻 | 예 (`{3.14159:...}`) |
|---|---|---|
| `:.2f` | 소수 둘째 자리까지 | `3.14` |
| `:,` | 천 단위 쉼표 | `{1234567:,}` → `1,234,567` |
| `:>8` | 오른쪽 정렬(8칸) | `    3.14` |
| `:03d` | 3자리, 빈 곳 0 채움 | `{7:03d}` → `007` |
| `:.0%` | 퍼센트 | `{0.75:.0%}` → `75%` |

````python
name = "지민"
score = 95
print(f"{name}님의 점수는 {score}점입니다")
print(f"5년 뒤 점수 예상: {score + 5}점")     # 중괄호 안에서 계산도 가능

pi = 3.141592
print(f"원주율: {pi:.2f}")               # 소수 둘째 자리 -> 3.14

label = "합계"
print(f"{1234567:,}원")                      # 천 단위 쉼표
print(f"[{label:>8}]")                         # 오른쪽 정렬 8칸
print(f"입장 번호: {7:03d}")                  # 007
print(f"진행률: {0.75:.0%}")                 # 75%
````

# 4. 연산자 (Operator)

## 4-1. 산술 연산자 — `/` 와 `//` 의 차이가 핵심!

| 연산자 | 뜻 | `7 ? 3` |
|---|---|---|
| `+ - *` | 더하기·빼기·곱하기 | `10`, `4`, `21` |
| `/` | 나누기 (**결과는 항상 실수**) | `2.3333...` |
| `//` | 몫 (소수점 **버림**) | `2` |
| `%` | 나머지 | `1` |
| `**` | 거듭제곱 | `2 ** 10 = 1024` |

````python
print(7 + 3)     # 10
print(7 - 3)     # 4
print(7 * 3)     # 21
print(7 / 3)     # 2.3333333333333335  <- / 는 항상 실수!
print(7 // 3)    # 2   <- 몫 (소수점 버림)
print(7 % 3)     # 1   <- 나머지
print(2 ** 10)   # 1024 <- 거듭제곱
````

> **`%`(나머지)의 쓸모**: `number % 2 == 0` 이면 짝수! 나중에 조건문에서 아주 자주 씁니다.
> **`//`(몫)의 쓸모**: 100분은 `100 // 60` = 1시간, `100 % 60` = 40분. (이따 변환기에서 써먹어요!)

## 4-2. 비교 연산자 — 결과는 `True` / `False`

| 연산자 | 뜻 |
|---|---|
| `==` | 같다 |
| `!=` | 다르다 |
| `>` `<` | 크다 / 작다 |
| `>=` `<=` | 크거나 같다 / 작거나 같다 |

````python
print(3 > 2)      # True
print(3 == 3)     # True  ('같다'는 = 이 아니라 == !)
print(3 != 4)     # True
print(5 <= 5)     # True
print(10 < 2)     # False
````

## 4-3. 논리 연산자 — `and` / `or` / `not`

여러 조건을 엮을 때 씁니다.

- `A and B` : **둘 다** 참이면 참
- `A or B` : **하나라도** 참이면 참
- `not A` : 참↔거짓 뒤집기

### 단축 평가(short-circuit)
- `and` 는 앞이 **False** 면, 뒤는 **보지도 않고** 바로 False.
- `or` 는 앞이 **True** 면, 뒤는 **보지도 않고** 바로 True.

````python
print(True and False)   # False
print(True or False)    # True
print(not True)         # False
````

````python
# 여러 조건 엮기
age = 25
print(age >= 20 and age < 30)   # 20대인가? -> True
````

````python
# 단축 평가: 앞이 False라 뒤(우산)는 확인조차 안 함
is_raining = False
has_umbrella = True
print(is_raining and has_umbrella)          # False
````

## 4-4. 복합 할당 연산자 — `+=`, `-=`, `*=`, `//=` ...

`score = score + 10` 을 짧게 `score += 10` 으로 쓸 수 있습니다. **누적**할 때 정말 자주 씁니다.

````python
score = 100
score += 10      # score = score + 10
print(score)     # 110
score -= 20      # 90
print(score)
score *= 2       # 180
print(score)
score //= 7      # 25 (몫)
print(score)
````

## 4-5. 문자열에 쓰는 연산 — `+`, `*`, `len()`, `in`

문자열도 `+` 로 잇고, `*` 로 반복할 수 있어요. 길이는 `len()`, 포함 여부는 `in`.

````python
greeting = "안녕" + "하세요"      # 이어붙이기
print(greeting)                    # 안녕하세요
print("하" * 5)                # 하하하하하  (반복)
print(len("파이썬"))            # 3  (글자 수)
print("파" in "파이썬")         # True  (포함되어 있나?)
print("z" in "파이썬")          # False
````
