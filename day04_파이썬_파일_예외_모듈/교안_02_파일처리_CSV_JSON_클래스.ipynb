# 1. 텍스트 파일 읽고 쓰기

## 왜 필요할까요?

**파일에 저장**하면 프로그램을 껐다 켜도 데이터가 남습니다. 이게 '영속화(저장)'예요.

## open() 과 파일 모드
파일은 `open(경로, 모드, encoding)` 으로 엽니다.

| 모드 | 뜻 | 파일이 있으면 | 파일이 없으면 |
|---|---|---|---|
| `"r"` | 읽기 (read) | 읽는다 | ❌ 에러 |
| `"w"` | 쓰기 (write) | **덮어쓴다** | 새로 만든다 |
| `"a"` | 추가 (append) | 뒤에 이어 쓴다 | 새로 만든다 |

- 한글이 깨지지 않게 **항상 `encoding="utf-8"`** 을 붙여요.
- **`with`** 문을 쓰면 파일을 자동으로 안전하게 닫아 줍니다. (강력 추천!)

> 파일을 못 찾는 `FileNotFoundError` 가 나면 십중팔구 경로·폴더 문제입니다. 현재 폴더가 궁금하면 `import os; print(os.getcwd())` 로 확인하세요.

# 2. CSV 파일 다루기

## CSV 가 뭐예요?
**C**omma **S**eparated **V**alues — 값을 **쉼표로 구분**한 표 데이터예요. 엑셀도 CSV로 저장할 수 있죠.
```
이름,국어,영어,수학
김민준,88,92,95
```
첫 줄은 보통 **헤더(제목 줄)**, 나머지는 **데이터 줄**입니다.

> 📝 **일상 비유** — 엑셀 표를 메모장으로도 열 수 있게 '쉼표로 칸을 나눈' 가장 단순한 표예요.

## 읽는 두 가지 방법
| 방법 | 각 줄을 무엇으로? | 접근 |
|---|---|---|
| `csv.reader` | **리스트** | `row[0]`, `row[1]` (순서로) |
| `csv.DictReader` | **딕셔너리** | `row["이름"]` (제목으로) ✅ 편함 |

````python
# csv.reader — 각 줄을 '리스트'로 읽기
import csv
with open("data/students.csv", encoding="utf-8") as f:
    reader = csv.reader(f)
    header = next(reader)              # 첫 줄(header)을 따로 꺼내기
    print("헤더:", header)
    for row in reader:              # 나머지 줄
        print(row)
````

````python
# csv.DictReader — 각 줄을 '딕셔너리'로 읽기 (제목으로 접근! 훨씬 편해요)
import csv
with open("data/students.csv", encoding="utf-8") as f:
    students = list(csv.DictReader(f))
````

````python
# 전원의 평균 출력
for s in students:
    avg = (int(s["국어"]) + int(s["영어"]) + int(s["수학"])) / 3
    print(f"{s['이름']}: 평균 {avg:.1f}점")
````

````python
# 1등 찾기 — 평균을 기준(key)으로 max
def calc_avg(s):
    return (int(s["국어"]) + int(s["영어"]) + int(s["수학"])) / 3

top_student = max(students, key=calc_avg)
print("\n 1등:", top_student["이름"], f"(평균 {calc_avg(top_student):.1f}점)")
````

````python
# csv.writer — 계산 결과를 새 CSV 로 저장하기
import csv, os
os.makedirs("output", exist_ok=True)

summary = [["이름", "평균"]]                 # 첫 줄은 header
for s in students:
    avg = (int(s["국어"]) + int(s["영어"]) + int(s["수학"])) / 3
    summary.append([s["이름"], round(avg, 1)])

# newline="" 은 CSV 저장 시 줄 사이에 빈 줄이 끼는 걸 막아 줘요 (csv.writer 관례)
with open("output/성적요약.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(summary)             # 리스트의 리스트를 한 번에 쓰기

print("저장 완료 → output/성적요약.csv  (총", len(summary) - 1, "명)")
````

# 3. JSON 파일 다루기

## JSON 은 파이썬 딕셔너리의 쌍둥이
**J**ava**S**cript **O**bject **N**otation — 웹과 앱에서 데이터를 주고받는 표준 형식이에요.
생김새가 파이썬 **딕셔너리·리스트와 거의 똑같아서** 배우기 쉽습니다.

```json
{ "이름": "민준", "나이": 20, "취미": ["코딩", "게임"] }
```

| 하는 일 | 함수 | 방향 |
|---|---|---|
| 파일 → 파이썬 | `json.load(f)` | 읽기 |
| 파이썬 → 파일 | `json.dump(obj, f)` | 쓰기 |
| 문자열 → 파이썬 | `json.loads(s)` | 파싱 |
| 파이썬 → 문자열 | `json.dumps(obj)` | 직렬화 |

> **`load`·`dump` 는 파일**, **`loads`·`dumps` 는 문자열**(끝의 `s` = string)을 다뤄요.
> 내일 배울 API 는 데이터를 **문자열**로 보내주기 때문에, 받은 문자열을 `json.loads` 로 딕셔너리로 바꿔 쓰게 됩니다.

- 한글이 `\uae40` 처럼 깨지지 않게 → **`ensure_ascii=False`**
- 사람이 읽기 좋게 들여쓰기 → **`indent=2`**

````python
# json.load — JSON 파일을 파이썬 딕셔너리로 읽기
import json
with open("data/menu.json", encoding="utf-8") as f:
    cafe = json.load(f)

print("☕", cafe["카페이름"], "(", cafe["영업시간"], ")")
print("-" * 30)
for menu in cafe["메뉴"]:
    status = "❌품절" if menu["품절"] else "✅판매중"
    print(f"{menu['이름']:<8} {menu['가격']:>5}원  {status}")
````

````python
# json.dump — 수정한 데이터를 새 JSON 파일로 저장
import json, os
os.makedirs("output", exist_ok=True)

# 모든 menu 가격을 500원씩 인상해 봅시다.
for menu in cafe["메뉴"]:
    menu["가격"] += 500

with open("output/menu_수정.json", "w", encoding="utf-8") as f:
    json.dump(cafe, f, ensure_ascii=False, indent=2)

print("가격 인상 후 저장 완료 → output/menu_수정.json")
print("아메리카노 새 가격:", cafe["메뉴"][0]["가격"], "원")
````

````python
# load/dump 는 '파일', loads/dumps 는 '문자열' 을 다뤄요 (끝의 s = string)
import json

# 파이썬 -> 문자열 (dumps)
menu = {"이름": "핫초코", "가격": 4800}
text = json.dumps(menu, ensure_ascii=False)
print("문자열로 :", text, "| 타입:", type(text).__name__)   # str

# 문자열 -> 파이썬 (loads)
back = json.loads(text)
print("다시 딕셔너리로:", back["이름"], "/", back["가격"], "원")
````

# 5. 클래스 — 데이터와 동작을 하나로 묶기

## 왜 필요할까요?
지금까지는 학생 정보를 딕셔너리 `{"name": "지민", "score": 85}` 로 담고, 그걸 다루는 동작(평균 내기 등)은 **따로 함수**로 만들었어요. 데이터와 동작이 **흩어져** 있죠.

**클래스(class)** 는 서로 관련된 **데이터(속성)** 와 **동작(메서드)** 을 **하나의 틀**로 묶습니다.

> 🥠 **일상 비유 — 붕어빵 틀** : `class` 는 붕어빵을 찍는 **틀(설계도)**, 그 틀로 찍어낸 붕어빵 하나하나가 **객체(object)** 예요. 틀은 하나지만, 그 틀로 팥 붕어빵·슈크림 붕어빵을 여러 개 만들 수 있죠. 틀 = 클래스, 붕어빵 = 객체입니다.

## 문법 — `class` · `__init__` · `self`
| 요소 | 뜻 |
|---|---|
| `class 이름:` | 새 틀(클래스)을 정의 |
| `__init__(self, ...)` | 객체를 **만들 때 자동 실행**되는 초기 설정 (이름·나이 등을 받아 저장) |
| `self` | **그 객체 자신** — 속성을 `self.name` 처럼 붙여 보관 |
| 속성(attribute) | 객체가 **가진 값** (`self.name`, `self.age`) |
| 메서드(method) | 클래스 안의 **함수** (첫 번째 인자는 항상 `self`) |

### 한눈에 보는 클래스 vs 객체

**클래스**는 붕어빵 **틀**(설계도) 하나, **객체**는 그 틀로 찍어낸 붕어빵 여러 개예요. 같은 틀에서 나와도 각자 다른 속성값(이름·나이)을 가집니다.

<img src="images/class_object.png" alt="클래스=틀, 객체=인스턴스" width="640">

````python
# 클래스 정의: 강아지를 나타내는 '틀'
class Dog:
    def __init__(self, name, age):    # 강아지를 만들 때 이름·나이를 받아요
        self.name = name              # self.name = 이 강아지의 이름 (속성)
        self.age = age

    def bark(self):                   # 메서드 = 클래스 안의 함수 (첫 인자는 self)
        print(f"{self.name}: 멍멍!")

# 틀(Dog)로 실제 강아지(객체)를 찍어낸다
# 같은 틀(Dog)로 서로 다른 강아지 3마리를 찍어낸다 (Dog("초코", 3) -> 이름=초코, 나이=3)
choco = Dog("초코", 3)
bori = Dog("보리", 5)
happy = Dog("해피", 1)

print(choco.name, choco.age, "살")   # 속성 꺼내기 -> 초코 3 살
choco.bark()                         # 메서드 호출 -> 초코: 멍멍!
bori.bark()                          # 보리: 멍멍!
happy.bark()                         # 해피: 멍멍!

# print(choco) 하면 <__main__.Dog object at 0x...> 처럼 낯선 게 나와요.
# 속성은 choco.name 처럼 하나씩 꺼내 보세요.
````

> 📌 **`self` 가 헷갈릴 때** — `choco.bark()` 라고 부르면, 파이썬이 **점(`.`) 앞의 `choco` 를 `bark(self)` 의 `self` 자리에 자동으로 넣어** 줍니다. 그래서 **정의할 땐 `self` 를 쓰지만, 부를 땐 적지 않아요.** `choco.bark()` 는 사실 `Dog.bark(choco)` 인 셈이고, 메서드 안의 `self.name` 은 곧 `choco.name` 이 됩니다.

### 메서드가 속성을 바꾸는 클래스
클래스의 진짜 힘은 **객체가 자기 상태(속성)를 기억하고, 메서드로 그걸 바꾸는** 데 있어요.
예금·출금으로 잔액이 변하는 통장을 만들어 봅시다.

````python
class BankAccount:
    def __init__(self, owner):
        self.owner = owner
        self.balance = 0              # 시작 잔액 0

    def deposit(self, amount):        # 입금 -> 잔액 증가
        self.balance += amount

    def withdraw(self, amount):       # 출금 -> 잔액 감소 (부족하면 안내)
        if amount > self.balance:
            print("잔액이 부족합니다!")
        else:
            self.balance -= amount

acc = BankAccount("민준")
acc.deposit(10000)
acc.withdraw(3000)
print(acc.owner, "잔액:", acc.balance, "원")   # 민준 잔액: 7000 원
acc.withdraw(999999)                          # 잔액이 부족합니다!
````

### 딕셔너리 vs 클래스
| | 딕셔너리 | 클래스 |
|---|---|---|
| 담는 것 | **데이터만** (`{"balance": 0}`) | **데이터 + 동작** 한 묶음 |
| 동작 | 함수를 따로 만들어 넘김 | 객체가 **자기 메서드**로 처리 (`acc.deposit(...)`) |
| 여러 개 | 딕셔너리를 여러 개 | **같은 틀**로 객체를 여러 개 (`Dog("초코")`, `Dog("보리")`) |

> 데이터가 몇 개 안 되면 딕셔너리로 충분하지만, **"이 데이터에 딸린 동작"** 이 많아지면 클래스로 묶는 게 훨씬 깔끔해요.
