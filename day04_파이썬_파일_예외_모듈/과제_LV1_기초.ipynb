## 9. 점수 검증기 (raise 로 잘못된 값 막기)
**배경**: 함수가 이상한 값을 받으면 **직접 에러를 일으켜** 알려 줄 수 있어요. 이게 `raise` 입니다.

**요구사항**:
- `set_score(score)` 함수를 만드세요.
- `score` 가 `0` 미만이거나 `100` 초과이면 `raise ValueError(...)` 로 에러를 일으킵니다.
- 정상 범위면 `score` 를 그대로 반환합니다.

**예시**
```
set_score(90)   ->  90
set_score(-1)   ->  ValueError 발생
set_score(101)  ->  ValueError 발생
```

<details><summary>힌트</summary>

```text
접근방법:
- 범위를 벗어나면 raise 로 에러를 던지고, 아니면 값을 그대로 돌려준다

세부구현:
1. score 가 0보다 작거나 100보다 크면 raise ValueError("...") 로 에러 발생

```

</details>

````python
# 여기에 코드를 작성하세요
````

````python
# [자가채점]
assert set_score(90) == 90
for bad in (-1, 101):
    try:
        set_score(bad)
        passed = False
    except ValueError:
        passed = True
    assert passed, f"{bad} 은 ValueError 를 일으켜야 해요"
print("✅ 문제9 통과!")
````

```

세부구현:
1. os.makedirs 로 상위 폴더를 먼저 만든다 (exist_ok=True)

```

````python
# [자가채점]
import os
log_path = "output/접속_test.log"
if os.path.exists(log_path):
    os.remove(log_path)              # 이전 실행 흔적을 지우고 깨끗하게 검증
append_log(log_path, "첫 번째 줄")
append_log(log_path, "두 번째 줄")
append_log(log_path, "세 번째 줄")
with open(log_path, encoding="utf-8") as f:
    lines = [ln for ln in f.read().splitlines() if ln]
assert len(lines) == 3, "세 번 호출하면 3줄이 쌓여야 해요 (a 모드)"
print("✅ 문제13 통과!")
````
