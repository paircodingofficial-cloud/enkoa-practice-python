# 1. 예외 처리 (Exception Handling)

## 왜 필요할까요?
지금까지 코드를 짜다 보면 빨간 글씨의 **에러**를 수도 없이 만났을 거예요.
에러가 나면 프로그램은 그 자리에서 **뚝 멈춰** 버립니다.

> **일상 비유** — 요리 중에 계란이 없다고 요리를 통째로 포기하나요?
> 보통은 "계란이 없네? 그럼 계란은 빼고 만들자" 하고 **대처**하죠.
> 예외 처리는 프로그램에게 이런 **대처 능력**을 주는 문법입니다.

## 에러 메시지(traceback) 읽는 법
에러가 나면 파이썬은 **traceback** 이라는 '사건 보고서'를 보여 줍니다. 겁먹지 말고 **맨 아래 줄부터** 읽으세요.

아래는 `int("사과")` 를 실행했을 때 나오는 traceback 입니다. 한 줄씩 해부해 봅시다.

```
Traceback (most recent call last):        ← "지금부터 에러 보고서예요"
  File "test.py", line 3, in <module>     ← 어디서? 파일 test.py 3번째 줄
    n = int("사과")                        ← 문제의 그 코드
ValueError: invalid literal for int()      ← ★ 무엇이? ValueError (핵심!)
     with base 10: '사과'                   ← 왜? '사과'는 정수로 못 바꿔요
```

핵심은 **맨 아랫줄**: `에러이름: 설명`.
- `ValueError` → "값(value)의 형태가 잘못됐다"
- 설명을 보면 `'사과'` 를 정수로 바꾸려다 실패했다는 걸 알 수 있죠.

**에러 이름만 알아도 절반은 해결**입니다.

먼저 **진짜 빨간 에러 화면**을 직접 한 번 봐 두면, 다음부터 겁이 덜 나요. 아래 셀에서 관찰해 봅시다.

````python
# [제공 코드] 진짜 에러 화면(traceback)을 눈으로 보고 싶다면, 아래 줄의 주석을 풀고 실행해 보세요.
# 빨간 글씨와 함께 셀이 '뚝' 멈추는 걸 관찰한 뒤, 다시 주석 처리하고 넘어가면 됩니다.
# print(int("삼"))   # <- 주석을 풀면 ValueError 가 나면서 프로그램이 멈춥니다!

print("위 줄의 주석을 풀어 에러 화면을 관찰해 보세요. 이제 이 에러를 try 로 잡아 봅니다. ")
````

````python
# 파이썬이 자주 던지는 대표 예외 6종을 일부러 만나 봅니다.
# 각 예외를 try/except 로 '잡아서' 어떤 상황에서 나는지 눈으로 확인해요.

# 1) ValueError — 값의 형태가 안 맞을 때
try:
    n = int("사과")            # 숫자가 아닌 문자열을 int 로
except ValueError as e:
    print("① ValueError:", e)
````

````python
# 2) TypeError — 자료형이 안 맞는 연산
try:
    sentence = "나이: " + 20        # 문자열 + 정수 (섞으면 안 됨)
except TypeError as e:
    print("② TypeError:", e)
````

````python
# 3) ZeroDivisionError — 0으로 나눔
try:
    x = 10 / 0
except ZeroDivisionError as e:
    print("③ ZeroDivisionError:", e)
````

````python
# 4) IndexError — 리스트 범위 밖 접근
try:
    data = [1, 2, 3]
    print(data[5])
except IndexError as e:
    print("④ IndexError:", e)
````

````python
# 5) KeyError — 딕셔너리에 없는 키
try:
    info = {"이름": "민준"}
    print(info["나이"])
except KeyError as e:
    print("⑤ KeyError: 키를", e, "를 찾을 수 없음")
````

````python
# 6) FileNotFoundError — 없는 파일 열기
try:
    open("없는파일.txt", encoding="utf-8")
except FileNotFoundError as e:
    print("⑥ FileNotFoundError:", e)
````

## try / except 문법
에러가 날 수 있는 코드를 `try:` 안에, 에러가 났을 때 할 일을 `except:` 안에 씁니다.

| 키워드 | 의미 | 언제 실행? |
|---|---|---|
| `try` | "이 코드는 위험할 수 있어" | 항상 먼저 시도 |
| `except` | "에러가 나면 이걸 해" | try 에서 에러가 났을 때만 |
| `else` | "에러가 **안** 났으면 이걸 해" | try 가 무사히 끝났을 때만 |
| `finally` | "무슨 일이 있어도 이건 해" | 에러 여부와 상관없이 **항상** |

> ⚠️ **try 를 남용하지 마세요.** 위험한 **줄만 골라서** 감싸세요. 코드 전체를 `try` 로 덮으면 엉뚱한 곳의 **진짜 버그까지 예외 뒤에 숨습니다.** 그리고 `except` 에는 **잡을 예외 타입을 콕 집어** 적으세요 — 그냥 `except:` 로 두면 모든 에러를 삼켜 버려 원인을 못 찾아요.

### 한눈에 보는 실행 흐름

`try` 를 실행하다 **에러가 나면 `except`**, **에러가 없으면 `else`**, 그리고 **`finally` 는 어느 쪽이든 항상** 실행됩니다.

<img src="images/try_except_flow.png" alt="try/except/else/finally 실행 흐름" width="380">

````python
# 가장 기본형: 위험한 코드를 try 로 감싸기
try:
    a = 10
    b = 0
    print(a / b)                 # 여기서 ZeroDivisionError 발생
except ZeroDivisionError:
    print("0으로는 나눌 수 없어요!")

print("→ 에러가 났지만 프로그램은 죽지 않고 계속됩니다.")
````

````python
# 예외 '타입 지정' + '다중 except': 상황마다 다르게 대응
def safe_division(a, b):
    try:
        return a / b
    except ZeroDivisionError:      # 0으로 나눈 경우
        print("[안내] 0으로 나눌 수 없습니다.")
    except TypeError:              # 숫자가 아닌 값이 들어온 경우
        print("[안내] 숫자만 넣어 주세요.")
    return None

print(safe_division(10, 2))     # 정상
print(safe_division(10, 0))     # ZeroDivisionError 처리
print(safe_division(10, "이"))  # TypeError 처리
````

````python
# else / finally: '성공' 과 '실패' 두 경우를 나란히 보기
def try_convert(text):
    try:
        num = int(text)
    except ValueError:
        print(f'  "{text}" -> 변환 실패 (except 실행, else 는 건너뜀)')
    else:
        print(f'  "{text}" -> 변환 성공! 값 = {num} (else 실행)')
    finally:
        print("  finally: 성공하든 실패하든 이 줄은 '항상' 실행돼요")

print("[성공 케이스]")
try_convert("42")
print("[실패 케이스]")
try_convert("사과")
````

## raise — 내가 직접 예외를 일으키기
가끔은 "이건 잘못된 값이야!" 하고 **내가 먼저 에러를 던져야** 할 때가 있어요.
예를 들어 나이에 `-5` 가 들어오면, 조용히 넘어가기보다 **분명하게 에러로 알리는** 편이 낫습니다.

````python
# raise: 규칙에 어긋나는 값이면 예외를 '일으켜' 알린다
def verify_age(age):
    if age < 0:
        raise ValueError("나이는 음수가 될 수 없어요")
    return f"{age}살이군요!"

# 호출하는 쪽에서 try 로 안전하게 받는다
try:
    print(verify_age(-5))
except ValueError as e:
    print("걸러냄:", e)

print(verify_age(20))   # 정상 값
````

> **참고 — 이름은 영어 snake_case로**
>
> 파이썬은 함수·변수 이름을 `safe_divide`, `total_price` 처럼 **영어 snake_case**(소문자+밑줄)로 짓는 것이 관례예요. 앞으로 쓸 외부 라이브러리도 모두 영어 이름이라 자연스럽게 어울립니다. (한글 설명·출력·데이터는 그대로 한글을 씁니다.)

# 2. 모듈 · 패키지 · 외부 라이브러리

## 왜 필요할까요?
필요한 기능을 **매번 직접 만들 필요는 없어요.** 누군가 이미 잘 만들어 둔 코드 묶음을 가져다 쓰면 됩니다.

> **일상 비유** — 요리할 때 소금을 직접 정제하지 않죠. 마트에서 사 옵니다.
> **모듈**은 '이미 만들어진 기능 상자'이고, **`import`** 는 그 상자를 '가져오는' 행동이에요.

- **모듈(module)**: 함수·변수가 든 `.py` 파일 하나 (예: `math`)
- **패키지(package)**: 모듈 여러 개를 담은 폴더 (예: `rich`)
- **표준 라이브러리**: 파이썬에 기본 내장 (설치 불필요 — `math`, `random`, `datetime` …)
- **외부 라이브러리**: 따로 설치해야 함 (`uv add rich` 처럼 — 오늘은 `rich`)

## import 하는 3가지 방법

````python
# 방법 1) 모듈 전체를 가져오기 ->  모듈이름.기능  으로 사용
import math
print("방법1:", math.sqrt(16))
````

````python
# 방법 2) 필요한 것만 콕 집어 가져오기 ->  기능  을 바로 사용
from math import pi, factorial
print("방법2:", pi, factorial(5))
````

````python
# 방법 3) 별명(as) 붙여 가져오기 ->  긴 이름을 짧게
import random as rnd
print("방법3:", rnd.randint(1, 100))
````

## 표준 라이브러리 투어
설치 없이 바로 쓸 수 있는 3대장을 만나 봅시다: **math**(수학), **random**(무작위), **datetime**(날짜·시간).

````python
# math — 수학 계산 도구 상자
import math
print("원주율 π :", math.pi)
print("√2      :", math.sqrt(2))
print("3의 4제곱:", math.pow(3, 4))
print("올림/내림:", math.ceil(3.2), "/", math.floor(3.8))
````

````python
# random — 무작위(랜덤) 도구 상자
import random
print("주사위   :", random.randint(1, 6))
menu = ["아메리카노", "라떼", "에이드"]
print("오늘의 추천:", random.choice(menu))
random.shuffle(menu)
print("메뉴 섞기 :", menu)
````

````python
# datetime — 날짜와 시간 도구 상자
from datetime import datetime, date, timedelta
now = datetime.now()
print("지금 시각:", now.strftime("%Y-%m-%d %H:%M"))
today = date.today()
print("오늘     :", today)
print("일주일 뒤:", today + timedelta(days=7))

# 특정 날짜를 직접 만들 수도 있어요: date(년, 월, 일)
christmas = date(2026, 12, 25)
print("크리스마스:", christmas)
print("100일 뒤 :", christmas + timedelta(days=100))
````

## 내가 만든 모듈 import 하기
모듈이 별거 아니라는 걸 눈으로 확인해 봅시다. **함수가 든 `.py` 파일**이 곧 모듈이에요.
아래 셀은 노트북에서 직접 `my_utils.py` 라는 파일을 만들어 냅니다. (실행 후 파일 탐색기에서 확인해 보세요!)

> ⚠️ **캐싱 주의**: 한 번 `import` 한 모듈은 파일 내용을 고쳐도 자동으로 반영되지 않습니다. 모듈 파일을 수정했다면 **커널을 재시작(Restart)** 한 뒤 다시 `import` 하세요.

````python
# [제공 코드] 노트북에서 직접 모듈 파일(my_utils.py)을 만들어 봅니다.
# open(..., "w") 로 파일에 함수 두 개를 써 넣어요. (파일 처리는 오후에 자세히!)
module_code = '''# my_utils.py — 내가 만든 첫 모듈
def greet(name):
    return f"안녕하세요, {name}님! 좋은 하루예요."

def area(radius):
    return 3.14159 * radius * radius
'''
with open("my_utils.py", "w", encoding="utf-8") as f:
    f.write(module_code)
print("my_utils.py 파일을 만들었어요! 이제 이걸 import 해 봅시다.")
````

````python
# [제공 코드] 방금 만든 모듈을 불러와 사용합니다.
import sys, os
sys.path.insert(0, os.getcwd())   # 현재 폴더를 파이썬이 찾도록 알려주기 (주피터는 보통 자동)

import my_utils                    # my_utils.py 를 모듈로 가져오기

print(my_utils.greet("도윤"))
print("반지름 5인 원의 넓이:", my_utils.area(5))
````

## 패키지 = 모듈을 담은 폴더
모듈이 많아지면 폴더로 묶어 정리합니다. 그게 **패키지**예요.

```
my_project/
├─ main.py              ← 실행 파일
├─ my_utils.py          ← 모듈 (함수 모음)
└─ shop/                ← 패키지 (폴더)
   ├─ __init__.py       ← "나는 패키지예요" 표시 파일
   ├─ menu.py           ← shop 안의 모듈
   └─ pay.py            ← shop 안의 모듈
```

이렇게 되면 `from shop import menu` 처럼 **폴더.모듈** 구조로 가져올 수 있어요.
우리가 곧 쓸 `rich` 도 이런 패키지랍니다.

## 그림 속 `shop/` 패키지를 진짜로 만들어 보기
위 그림은 눈으로만 봤죠. 이제 **그 구조를 실제 폴더·파일로 만들고, 불러와 실행**까지 해 봅니다.

````python
# [제공 코드] 그림과 똑같은 shop 패키지를 진짜 폴더·파일로 만듭니다.
import os
os.makedirs("shop", exist_ok=True)

menu_py = """MENU = {"아메리카노": 3000, "카페라떼": 3500, "레모네이드": 4000}

def show_menu():
    for name, price in MENU.items():
        print(f"{name}: {price}원")
"""

pay_py = """def calc_change(price, paid):
    # 거스름돈 = 낸 돈 - 가격
    return paid - price

def apply_discount(price, rate):
    # rate 는 0~1 사이 할인율 (0.1 이면 10% 할인)
    return int(price * (1 - rate))
"""

init_py = "# 이 파일이 있으면 shop 폴더가 '패키지' 가 됩니다. (지금은 비어 있어요)\n"

for name, body in [("menu.py", menu_py), ("pay.py", pay_py), ("__init__.py", init_py)]:
    with open(os.path.join("shop", name), "w", encoding="utf-8") as f:
        f.write(body)

print("shop/ 폴더 안 파일들:", os.listdir("shop"))   # 진짜로 생겼는지 눈으로 확인!
````

## 만든 패키지를 3가지 방법으로 불러와 실행하기
그림의 "`from shop import menu` 처럼 폴더.모듈 구조" 를 실제 코드로 확인해 봅시다.

````python
# 3가지 import 형태를 실제로 실행해 비교합니다.
import sys, os
sys.path.insert(0, os.getcwd())   # 현재 폴더를 파이썬이 찾도록 (주피터는 보통 자동)

# ① 폴더.모듈 통째로 가져오기 -> shop.menu.show_menu()
import shop.menu
shop.menu.show_menu()

print("-" * 20)
````

````python
# ② 패키지에서 모듈만 가져오기 -> pay.calc_change(...)
from shop import pay
print("거스름돈:", pay.calc_change(3000, 5000))

print("-" * 20)
````

````python
# ③ 모듈에서 함수만 콕 집어 가져오기 -> calc_change(...)
from shop.pay import calc_change
print("거스름돈:", calc_change(4000, 5000))
````

## `__init__.py` 를 채우면? — 재노출과 `__all__`
지금 `shop/__init__.py` 는 비어 있어서 `shop.menu.show_menu` 처럼 **모듈 이름(menu)까지** 적어야 했어요.
`__init__.py` 에 자주 쓰는 함수를 미리 불러 두면 `shop.show_menu` 처럼 **짧게** 쓸 수 있습니다. 직접 전/후를 비교해 봅시다.

> 🔎 **여기부터는 심화예요.** 어려우면 "이렇게 짧게 쓰는 방법도 있구나" 정도만 알고 넘어가도 됩니다.

````python
# (전) __init__.py 가 비어 있을 때: from shop import * 로는 함수가 바로 안 들어와요
from shop import *
print("show_menu 바로 사용 가능?", "show_menu" in dir())   # 비어 있어서 -> False
````

````python
# [제공 코드] __init__.py 를 '재노출 + __all__' 버전으로 다시 씁니다.
new_init = """from .menu import show_menu       # shop.show_menu 로 바로 쓰게 재노출 (.menu = 같은 패키지)
from .pay import calc_change

__all__ = ["show_menu", "calc_change"]   # from shop import * 로 공개할 이름 목록
"""
with open("shop/__init__.py", "w", encoding="utf-8") as f:
    f.write(new_init)

# 이미 import 한 패키지는 파일을 고쳐도 자동 반영되지 않아요(모듈 캐싱).
# 실습에서는 importlib.reload 로 다시 읽습니다. (평소엔 커널 재시작이 가장 확실해요.)
import importlib, shop
importlib.reload(shop)

print("이제 shop.show_menu 바로 사용 가능?", hasattr(shop, "show_menu"))
print("공개 목록(__all__):", shop.__all__)
````

````python
# (후) __init__.py 를 채운 뒤: 이제 함수가 바로 들어옵니다
from shop import *
print("show_menu 바로 사용 가능?", "show_menu" in dir())      # __all__ 에 있음 -> True
print("calc_change 바로 사용 가능?", "calc_change" in dir())  # __all__ 에 있음 -> True
show_menu()   # shop. 없이 바로 호출!
````

## 모듈 vs 패키지 한눈에

| 구분 | 모듈 (module) | 패키지 (package) |
|---|---|---|
| 정체 | 함수가 든 `.py` 파일 하나 | 모듈을 담은 **폴더** |
| 표식 | (없음) | 폴더 안 `__init__.py` |
| 예 | `my_utils.py` | `shop/`, `rich` |
| 가져오기 | `import my_utils` | `import shop.menu` / `from shop import show_menu` |

## 외부 라이브러리 — uv 로 설치하고 rich 로 예쁘게 출력하기
표준 라이브러리에 없는 기능은 **외부 라이브러리**를 설치해 씁니다. 우리는 환경 구성 때 배운 **uv** 로 관리해요.

터미널(코드 셀 아님!)에서 이렇게 설치합니다
```bash
uv add rich
```
> 이 실습 환경에는 `rich` 가 **이미 설치되어 있어요.** 위 명령은 새 라이브러리를 추가할 때의 방법으로 기억해 두세요.
> (⚠️ `uv add ...` 같은 터미널 명령은 코드 셀에 쓰면 안 됩니다. 반드시 터미널에서 실행해요.)

이제 `rich` 로 표와 색깔을 출력해 봅시다.

````python
# rich — 터미널을 예쁘게! (table, 색깔 출력)
from rich.console import Console
from rich.table import Table

console = Console()

table = Table(title="오늘의 학습 메뉴")
table.add_column("순서", justify="center")
table.add_column("주제", style="cyan")
table.add_column("난이도", justify="center")
table.add_row("1", "예외 처리", "★★☆")
table.add_row("2", "모듈 · 패키지", "★★☆")
table.add_row("3", "외부 라이브러리 (rich)", "★☆☆")
console.print(table)

console.print("[bold green]잘하고 있어요![/bold green] rich 로 [yellow]색깔 출력[/yellow]까지 성공했습니다.")
````

# 이번 강의 정리

| 주제 | 핵심 한 줄 |
|---|---|
| 에러 읽기 | traceback 은 **맨 아랫줄**(`에러이름: 설명`)부터 읽는다 |
| try / except | 위험한 코드를 감싸 프로그램이 죽지 않게 |
| 다중 except | 예외 종류마다 다르게 대응 |
| else / finally | 성공 시 / 무조건 실행할 일 |
| raise | 잘못된 값을 내가 먼저 에러로 알리기 |
| import | `import` / `from...import` / `import...as` 3가지 |
| 표준 라이브러리 | `math` · `random` · `datetime` (설치 불필요) |
| 내 모듈 | 함수가 든 `.py` 파일이 곧 모듈 |
| 외부 라이브러리 | `uv add 이름` 으로 설치 (오늘은 `rich`) |
