# 1. 텍스트 파일 읽고 쓰기

## 왜 필요할까요?
지금까지 만든 데이터는 프로그램이 끝나면 **모두 사라졌어요.** 변수는 메모리에만 있으니까요.
**파일에 저장**하면 프로그램을 껐다 켜도 데이터가 남습니다. 이게 '영속화(저장)'예요.

> 📝 **일상 비유** — 머릿속 생각(변수)은 잊혀지지만, 공책에 적으면(파일) 나중에 다시 봅니다.

## open() 과 파일 모드
파일은 `open(경로, 모드, encoding)` 으로 엽니다.

| 모드 | 뜻 | 파일이 있으면 | 파일이 없으면 |
|---|---|---|---|
| `"r"` | 읽기 (read) | 읽는다 | ❌ 에러 |
| `"w"` | 쓰기 (write) | **덮어쓴다** | 새로 만든다 |
| `"a"` | 추가 (append) | 뒤에 이어 쓴다 | 새로 만든다 |

- 한글이 깨지지 않게 **항상 `encoding="utf-8"`** 을 붙여요.
- **`with`** 문을 쓰면 파일을 자동으로 안전하게 닫아 줍니다. (강력 추천!)

> **경로는 '이 노트북이 있는 폴더' 기준(상대경로)이에요.** `data/students.csv` 는 "지금 이 노트북 옆의 `data` 폴더 안 파일"을 뜻해요.
> 파일을 못 찾는 `FileNotFoundError` 가 나면 십중팔구 경로·폴더 문제입니다. 현재 폴더가 궁금하면 `import os; print(os.getcwd())` 로 확인하세요.

# 2. CSV 파일 다루기

## CSV 가 뭐예요?
**C**omma **S**eparated **V**alues — 값을 **쉼표로 구분**한 표 데이터예요. 엑셀도 CSV로 저장할 수 있죠.
```
이름,국어,영어,수학
김민준,88,92,95
```
첫 줄은 보통 **헤더(제목 줄)**, 나머지는 **데이터 줄**입니다.

> 📝 **일상 비유** — 엑셀 표를 메모장으로도 열 수 있게 '쉼표로 칸을 나눈' 가장 단순한 표예요.

## 읽는 두 가지 방법
| 방법 | 각 줄을 무엇으로? | 접근 |
|---|---|---|
| `csv.reader` | **리스트** | `row[0]`, `row[1]` (순서로) |
| `csv.DictReader` | **딕셔너리** | `row["이름"]` (제목으로) ✅ 편함 |

````python
# csv.reader — 각 줄을 '리스트'로 읽기
import csv
with open("data/students.csv", encoding="utf-8") as f:
    reader = csv.reader(f)
    header = next(reader)              # 첫 줄(header)을 따로 꺼내기
    print("헤더:", header)
    for row in reader:              # 나머지 줄
        print(row)
````

````python
# csv.DictReader — 각 줄을 '딕셔너리'로 읽기 (제목으로 접근! 훨씬 편해요)
import csv
with open("data/students.csv", encoding="utf-8") as f:
    students = list(csv.DictReader(f))
````

````python
# 전원의 평균 출력
for s in students:
    avg = (int(s["국어"]) + int(s["영어"]) + int(s["수학"])) / 3
    print(f"{s['이름']}: 평균 {avg:.1f}점")
````

````python
# 1등 찾기 — 평균을 기준(key)으로 max
def calc_avg(s):
    return (int(s["국어"]) + int(s["영어"]) + int(s["수학"])) / 3

top_student = max(students, key=calc_avg)
print("\n 1등:", top_student["이름"], f"(평균 {calc_avg(top_student):.1f}점)")
````

````python
# csv.writer — 계산 결과를 새 CSV 로 저장하기
import csv, os
os.makedirs("output", exist_ok=True)

summary = [["이름", "평균"]]                 # 첫 줄은 header
for s in students:
    avg = (int(s["국어"]) + int(s["영어"]) + int(s["수학"])) / 3
    summary.append([s["이름"], round(avg, 1)])

# newline="" 은 CSV 저장 시 줄 사이에 빈 줄이 끼는 걸 막아 줘요 (csv.writer 관례)
with open("output/성적요약.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(summary)             # 리스트의 리스트를 한 번에 쓰기

print("저장 완료 → output/성적요약.csv  (총", len(summary) - 1, "명)")
````

# 3. JSON 파일 다루기

## JSON 은 파이썬 딕셔너리의 쌍둥이
**J**ava**S**cript **O**bject **N**otation — 웹과 앱에서 데이터를 주고받는 표준 형식이에요.
생김새가 파이썬 **딕셔너리·리스트와 거의 똑같아서** 배우기 쉽습니다.

```json
{ "이름": "민준", "나이": 20, "취미": ["코딩", "게임"] }
```

| 하는 일 | 함수 | 방향 |
|---|---|---|
| 파일 → 파이썬 | `json.load(f)` | 읽기 |
| 파이썬 → 파일 | `json.dump(obj, f)` | 쓰기 |
| 문자열 → 파이썬 | `json.loads(s)` | 파싱 |
| 파이썬 → 문자열 | `json.dumps(obj)` | 직렬화 |

> **`load`·`dump` 는 파일**, **`loads`·`dumps` 는 문자열**(끝의 `s` = string)을 다뤄요.
> 내일 배울 API 는 데이터를 **문자열**로 보내주기 때문에, 받은 문자열을 `json.loads` 로 딕셔너리로 바꿔 쓰게 됩니다.

- 한글이 `\uae40` 처럼 깨지지 않게 → **`ensure_ascii=False`**
- 사람이 읽기 좋게 들여쓰기 → **`indent=2`**

````python
# json.load — JSON 파일을 파이썬 딕셔너리로 읽기
import json
with open("data/menu.json", encoding="utf-8") as f:
    cafe = json.load(f)

print("☕", cafe["카페이름"], "(", cafe["영업시간"], ")")
print("-" * 30)
for menu in cafe["메뉴"]:
    status = "❌품절" if menu["품절"] else "✅판매중"
    print(f"{menu['이름']:<8} {menu['가격']:>5}원  {status}")
````

````python
# json.dump — 수정한 데이터를 새 JSON 파일로 저장
import json, os
os.makedirs("output", exist_ok=True)

# 모든 menu 가격을 500원씩 인상해 봅시다.
for menu in cafe["메뉴"]:
    menu["가격"] += 500

with open("output/menu_수정.json", "w", encoding="utf-8") as f:
    json.dump(cafe, f, ensure_ascii=False, indent=2)

print("가격 인상 후 저장 완료 → output/menu_수정.json")
print("아메리카노 새 가격:", cafe["메뉴"][0]["가격"], "원")
````

````python
# load/dump 는 '파일', loads/dumps 는 '문자열' 을 다뤄요 (끝의 s = string)
import json

# 파이썬 -> 문자열 (dumps)
menu = {"이름": "핫초코", "가격": 4800}
text = json.dumps(menu, ensure_ascii=False)
print("문자열로 :", text, "| 타입:", type(text).__name__)   # str

# 문자열 -> 파이썬 (loads)
back = json.loads(text)
print("다시 딕셔너리로:", back["이름"], "/", back["가격"], "원")
````

# 5. AI 활용 코드 리팩토링

## 리팩토링이 뭐예요?
**동작은 그대로 두고, 코드를 더 깔끔하게 고치는 것**입니다. (기능 추가가 아니에요!)
환경 구성 때 배운 **Codex** 같은 AI 도구는 리팩토링을 아주 잘 도와줍니다.

먼저 '고칠 거리가 많은' 코드를 봅시다. 돌아가긴 하지만 문제가 보이나요?

````python
# 리팩토링 '전' 코드 — 돌아가긴 하지만 문제가 많아요.
def print_scores():
    kor1 = 90
    eng1 = 80
    math1 = 70
    total1 = kor1 + eng1 + math1
    print("학생1 평균:", total1 / 3)

    kor2 = 60
    eng2 = 75
    math2 = 85
    total2 = kor2 + eng2 + math2
    print("학생2 평균:", total2 / 3)

    kor3 = 100
    eng3 = 95
    math3 = 90
    total3 = kor3 + eng3 + math3
    print("학생3 평균:", total3 / 3)

print_scores()
````

## 이 코드의 문제점
- **중복**: 똑같은 '평균 내는 로직'이 3번 반복돼요.
- **하드코딩된 숫자**: `/ 3` 의 `3` 이 뜬금없이 등장 (과목이 4개가 되면?) — 이런 숫자를 실무에서는 '매직 넘버'라고도 부릅니다.
- **긴 함수**: 한 함수가 너무 많은 일을 해요.

## AI에게 리팩토링을 부탁하는 프롬프트 3패턴
좋은 요청 = **역할 + 제약 + 단계 요청**. 아래 3가지 패턴을 기억하세요.

**패턴 ① 역할 부여**
> "너는 파이썬 강사야. 아래 코드를 **초보자도 이해할 수 있게** 리팩토링해 줘."

**패턴 ② 제약 조건 명시**
> "**함수로 중복을 없애고, 하드코딩된 숫자를 변수로** 빼 줘. 단, **동작(출력)은 똑같이** 유지해."

**패턴 ③ 단계별 설명 요청**
> "리팩토링한 코드와 함께, **무엇을 왜 바꿨는지 한 줄씩** 설명해 줘."

이 프롬프트로 요청하면 아래와 같은 결과를 받을 수 있어요.

````python
# 리팩토링 '후' — 중복 제거 + 함수 분리 + 하드코딩 제거
def get_avg(scores):
    return sum(scores) / len(scores)      # 과목이 몇 개든 OK (하드코딩 제거!)

scores_by_student = {
    "학생1": [90, 80, 70],
    "학생2": [60, 75, 85],
    "학생3": [100, 95, 90],
}

for name, score in scores_by_student.items():
    print(f"{name} 평균: {get_avg(score):.1f}")
````
