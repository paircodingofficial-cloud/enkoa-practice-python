## 1. 문자열 인덱싱 · 슬라이싱 (리스트 워밍업)

문자열은 **문자들이 순서대로 늘어선** 것입니다. 각 문자엔 **0부터** 시작하는 번호(인덱스)가 있어요.

```
 P  Y  T  H  O  N
 0  1  2  3  4  5     ← 앞에서부터
-6 -5 -4 -3 -2 -1     ← 뒤에서부터(음수)
```

>  **비유**: 사물함이 일렬로 있고 각 칸에 번호가 붙어 있는 모습.

**슬라이싱** `문자열[시작:끝:간격]` — 시작은 포함, **끝은 미포함**.

| 표현 | 의미 |
|---|---|
| `s[0]` | 0번(첫) 글자 |
| `s[-1]` | 맨 뒤 글자 |
| `s[0:3]` | 0,1,2번 (3은 제외) |
| `s[2:]` | 2번부터 끝까지 |
| `s[:3]` | 처음부터 2번까지 |
| `s[::-1]` | 거꾸로 뒤집기 |

````python
text = "PYTHON"
print(text[0], text[1], text[5])   # P Y N
print(text[-1], text[-2])         # N O
print("앞 3글자:", text[0:3])   # PYT
print("2번부터 :", text[2:])    # THON
print("처음~2번:", text[:3])    # PYT
print("거꾸로  :", text[::-1])  # NOHTYP
print("글자 수 :", len(text))   # 6
````

### 자주 쓰는 문자열 메서드

**함수와 메서드는 뭐가 다를까요?** 둘 다 "무언가를 해 주는 도구"지만, **부르는 방식**이 다릅니다.

| | 부르는 법 | 예 |
|---|---|---|
| **함수** | 이름을 부르고 **괄호 안에 재료**를 넣음 | `len(text)` · `print(text)` |
| **메서드** | 값 뒤에 **점(`.`)** 을 찍고 부름 | `text.upper()` · `text.strip()` |

- **함수**는 혼자 서 있는 도구입니다. `len()` 은 문자열·리스트 등 **여러 종류에 두루** 쓰여서 따로 있어요.
- **메서드**는 **그 값에 딸린** 도구입니다. "대문자로 바꾸기"는 문자열만 할 수 있는 일이라, 문자열 뒤에 붙여 `text.upper()` 로 씁니다.
- 그래서 재료의 위치가 반대로 보입니다 — `len(text)` 는 **괄호 안**에, `text.upper()` 는 **점 앞**에 재료가 옵니다.

> 리스트에는 리스트의 메서드(`.append()`), 딕셔너리에는 딕셔너리의 메서드(`.keys()`)가 따로 있습니다.
> **점을 찍었을 때 나오는 목록이 곧 "이 값이 할 수 있는 일"** 이에요.

````python
sentence = "  Hello, Python World  "
print("strip  :", sentence.strip())               # 양끝 공백 제거
print("upper  :", sentence.strip().upper())        # 대문자로
print("lower  :", sentence.strip().lower())        # 소문자로
print("replace:", sentence.replace("Python", "Java").strip())
print("split  :", "사과,바나나,포도".split(","))  # 리스트로 쪼개기
print("join   :", "-".join(["2026", "07", "10"]))  # 하나로 합치기
print("count  :", "banana".count("a"))          # a 개수 = 3
print("index  :", "python".index("t"))          # t의 위치 = 2
````

### 문자와 숫자 사이 — 아스키(ASCII) 코드

컴퓨터는 글자도 결국 **숫자로** 저장합니다.
"어떤 글자에 몇 번을 줄지" 미리 약속해 둔 표가 **아스키 코드표** 예요.

- `ord("문자")` -> 그 문자의 **번호**
- `chr(번호)` -> 그 번호의 **문자**

표에서 **알파벳** 칸을 보세요. 대문자 `A`는 **65**, 소문자 `a`는 **97** 입니다.
`A`~`Z`, `a`~`z` 가 각각 **26칸 연속으로** 붙어 있어서, 시작 번호에 하나씩 더하면 알파벳 전체를 만들 수 있습니다.

<img src="images/아스키_코드표.png" alt="아스키 코드표" width="588">

````python
# ord / chr : 문자 <-> 코드 번호 (시저 암호 등에 활용)
print("A의 코드:", ord("A"))   # 65
print("a의 코드:", ord("a"))   # 97
print("65번 문자:", chr(65))    # A
print("97번 문자:", chr(97))    # a
````

````python
# 반복문으로 알파벳 26글자 만들기 — 시작 문자의 코드에서 하나씩 더한다
for i in range(26):
    print(chr(ord("A") + i), end=" ")
print()

for i in range(26):
    print(chr(ord("a") + i), end=" ")
print()
````

## 2. 리스트(List) — 여러 값을 순서대로 담는 그릇

**대괄호 `[]`** 로 만듭니다. 값들을 **순서대로** 담고, **바꿀 수 있습니다**(가변).

>  **비유**: 장바구니. 물건을 순서대로 담고, 넣고 빼고 정렬할 수 있죠.

문자열에서 쓰던 **인덱싱 · 슬라이싱 · `len()` · `in`** 이 리스트에도 그대로 통합니다.

| 문법 | 하는 일 | 결과 (`fruits = ["사과", "바나나", "포도"]`) |
|---|---|---|
| `fruits[0]` | 인덱싱 — 그 자리의 값 **하나**를 꺼냄 | `"사과"` |
| `fruits[0:2]` | 슬라이싱 — 구간을 잘라 **새 리스트**로 | `["사과", "바나나"]` |
| `len(fruits)` | **길이** — 담긴 값이 몇 개인지 **숫자**로 | `3` |
| `"사과" in fruits` | **포함 여부** — 그 값이 있는지 **`True`/`False`** 로 | `True` |

- **`len()`** 은 개수를 세어 주는 함수입니다. 빈 리스트 `[]` 는 `0`, 그리고 **마지막 칸의 번호는 언제나 `len() - 1`** 이에요.
  (칸 번호가 0부터 시작하니, 3개짜리 리스트의 마지막은 `fruits[2]` 입니다.)
- **`in`** 은 위치를 알려 주는 게 아니라 **"있냐?" 만 묻습니다** → `True` 또는 `False`.
  몇 번째인지가 궁금하면 `.index()`, 있는지만 궁금하면 `in` 을 쓰세요. 반대는 `not in` 입니다.

**언제 쓰나요?** 같은 종류의 값이 **여러 개**이고, **순서가 있고**, **개수가 늘거나 줄 때**.
- 장바구니에 담은 상품들, 반 학생 명단, 하루치 기온 기록
- 파일에서 한 줄씩 읽어 차곡차곡 쌓기, 반복문으로 하나씩 처리하기
- 실무에서 **표의 행 목록**(엑셀 한 줄 = 값 하나)은 거의 다 리스트입니다.

````python
fruits = ["사과", "바나나", "포도"]
print(fruits)
print("첫번째:", fruits[0], "/ 마지막:", fruits[-1])
print("앞 2개:", fruits[0:2])
print("개수:", len(fruits))
print("사과 있나요?", "사과" in fruits)   # True
````

````python
# 리스트 메서드 : 추가 · 삭제 · 정렬
nums = [3, 1, 4, 1, 5]
nums.append(9)          # 끝에 추가
print("append :", nums)

nums.insert(1, 100)     # 1번 위치에 삽입
print("insert :", nums)

nums.remove(1)          # 값 1을 '첫 번째'만 제거
print("remove :", nums)

popped = nums.pop()      # 마지막 값을 꺼내며 제거
print("pop    :", popped, "→", nums)

nums.sort()             # 오름차순 정렬 (원본이 바뀜)
print("sort   :", nums)

nums.reverse()          # 순서 뒤집기
print("reverse:", nums)

print("index(4):", nums.index(4))       # 값 4의 위치
print("count(1):", [1, 1, 2, 1].count(1))  # 값 1의 개수 = 3
````

````python
# 리스트 순회 (어제 배운 for 그대로!)
for fruit in fruits:
    print("-", fruit)
````

### ⚠️ 리스트 복사할 때 조심할 것

`b = a` 는 리스트를 **복사하지 않습니다.** 같은 리스트에 **이름표를 하나 더 붙일** 뿐이에요.
그래서 `b` 를 바꾸면 `a` 도 같이 바뀝니다 — 초보자가 가장 많이 겪는 버그입니다.

진짜 복사본이 필요하면 **`.copy()`** 를 쓰세요.

````python
# b = a 는 복사가 아니라 '같은 리스트에 이름 하나 더 붙이기'
a = [1, 2, 3]
b = a
b.append(99)

print("a:", a)      # [1, 2, 3, 99]  <- 건드린 적 없는 a 까지 바뀐다!
print("b:", b)      # [1, 2, 3, 99]
````

````python
# .copy() 로 진짜 복사본 만들기 — 이제 서로 영향을 주지 않는다
c = [1, 2, 3]
d = c.copy()
d.append(99)

print("c:", c)      # [1, 2, 3]      <- 원본은 그대로
print("d:", d)      # [1, 2, 3, 99]
````

> 딕셔너리도 똑같습니다 — `d2 = d1` 은 이름만 하나 더 붙는 것이라 `.copy()` 가 필요해요.
> 반대로 **튜플·문자열은 애초에 바꿀 수 없으니** 이 걱정을 할 필요가 없습니다.

## 3. 튜플(Tuple) — 바꿀 수 없는 값 묶음

**소괄호 `()`** 로 만듭니다. 리스트와 비슷하지만 **한 번 정하면 바꿀 수 없습니다**(불변).

>  **비유**: 좌표 `(위도, 경도)`, 생년월일 `(년, 월, 일)` 처럼 **함부로 바뀌면 안 되는** 값 묶음.

그리고 앞서 궁금했던 것! **여러 값 반환의 정체가 바로 튜플**입니다.

인덱싱·슬라이싱·`len()`·`in` 은 **리스트와 똑같이** 됩니다. 못 하는 건 **값을 바꾸는 것** 하나뿐이에요.

**언제 쓰나요?** **한 세트로 묶여 다니고, 도중에 바뀌면 안 되는** 값일 때.
- 좌표 `(37.5, 127.0)`, 색 `(255, 0, 0)`, 이미지 크기 `(가로, 세로)`
- 함수가 **여러 값을 한 번에 반환**할 때 — 오늘 배운 바로 그것!
- **딕셔너리의 키로 쓸 수 있습니다** (리스트는 안 됩니다).
  절대 안 바뀐다고 보장되니 값을 찾는 **"주소"** 로 삼을 수 있거든요.
  예) `sales[(2026, 7)] = 1200` 처럼 `(년, 월)` 한 쌍을 키로 쓰기.

> **주의**: 값이 하나뿐인 튜플은 `(1,)` 처럼 **쉼표를 꼭** 찍어야 합니다.
> `(1)` 은 그냥 괄호로 감싼 숫자 `1` 이에요.

````python
point = (37.5, 127.0)
print("위도:", point[0], "/ 경도:", point[1])
# point[0] = 0    # 주석을 풀면 에러! 튜플은 한 번 만들면 수정 불가
````

````python
# 인덱싱·슬라이싱·len()·in - 리스트에서 하던 것이 그대로 됩니다
scores = (90, 85, 95, 70)
print("첫 점수:", scores[0], "/ 마지막:", scores[-1])
print("앞 2개 :", scores[0:2])        # 잘라낸 결과도 튜플
print("개수   :", len(scores))
print("95 있나?", 95 in scores)
````

````python
# 언패킹: 튜플의 값을 여러 변수로 한 번에 풀기
x, y = point
print("x =", x, ", y =", y)
````

````python
# 두 변수 교환 (튜플의 마법!)
a, b = 10, 20
a, b = b, a
print("교환 후 a =", a, ", b =", b)
````

````python
# 함수에서 본 '다중 반환' 의 정체 공개!
def quotient_remainder(a, b):
    return a // b, a % b     # 콤마로 반환 -> 사실 튜플로 묶여서 나옴

result = quotient_remainder(17, 5)
print("반환값:", result)
print("정체:", type(result))    # <class 'tuple'>
print("따로 꺼내기:", result[0], result[1])
````

## 4. 딕셔너리(Dict) — 이름표(키)로 값을 찾는 그릇

**중괄호 `{키: 값}`** 로 만듭니다. 순서 번호(인덱스) 대신 **키(key)** 로 값을 바로 찾습니다.

>  **비유**: 국어사전(단어→뜻), 전화번호부(이름→번호). "3번째" 가 아니라 **"이름으로"** 찾죠.

| 동작 | 문법 |
|---|---|
| 조회 | `d[키]` 또는 `d.get(키)` |
| 수정/추가 | `d[키] = 값` |
| 삭제 | `del d[키]` |
| 키 존재 확인 | `키 in d` |
| 전체 순회 | `d.keys()` · `d.values()` · `d.items()` |

**언제 쓰나요?** 값마다 **이름표**가 필요하고, 그 이름으로 **바로 찾아야** 할 때.
- 한 사람의 정보 `{"이름": "지민", "점수": 85}` — 0번·1번이 아니라 **뜻으로** 접근
- **API 응답과 설정 파일의 JSON 이 곧 딕셔너리**입니다. 앞으로 가장 자주 만날 구조예요.
- 개수 세기(단어 빈도), 코드↔이름 변환표
- 리스트에서 `in` 으로 찾으면 앞에서부터 훑지만, 딕셔너리는 **키로 곧장** 찾아 훨씬 빠릅니다.

````python
student = {"이름": "지민", "나이": 25, "전공": "AI"}
print("이름:", student["이름"])            # 대괄호 조회
print("나이:", student.get("나이"))        # get 조회
print("취미:", student.get("취미", "없음"))  # 없는 키는 기본값 반환 (에러 X)

student["나이"] = 26        # 수정 (있는 키)
student["학년"] = 3         # 추가 (없는 키에 대입)
print("수정·추가 후:", student)

del student["전공"]         # 삭제
print("삭제 후:", student)
print("'이름' 키 있나요?", "이름" in student)
````

````python
# keys / values / items 로 순회
scores = {"국어": 90, "영어": 85, "수학": 95}

print("과목들:", list(scores.keys()))
print("점수들:", list(scores.values()))

for key, val in scores.items():        # 키와 값을 한 번에!
    print(f"{key}: {val}점")
````

````python
# sum() 은 리스트·튜플 등 여러 숫자의 '합'을 구해 주는 내장 함수입니다
print("총점:", sum(scores.values()))
print("평균:", sum(scores.values()) / len(scores))
````

## 5. 집합(set) — 중복 없는 모음 (간단히)

**중괄호 `{}`** 에 값만 넣습니다(키:값 아님). **중복이 자동 제거**되고 순서가 없습니다.

**언제 쓰나요?** **중복을 없앨 때**, 그리고 **"있냐?" 만 빠르게 물을 때**.
- 중복 제거 — 이미 방문한 URL, 이미 처리한 주문 번호
- 두 모음 비교 — 두 과목을 함께 듣는 학생(교집합), 아직 안 본 항목(차집합)
- 순서가 없고 중복도 못 담으니, **순서가 중요하면 리스트**를 쓰세요.

````python
raw_nums = [1, 2, 2, 3, 3, 3]
unique = set(raw_nums)          # 중복 제거
print("중복 제거:", unique)

a = {1, 2, 3}
b = {2, 3, 4}
print("교집합(&):", a & b)   # {2, 3}
print("합집합(|):", a | b)   # {1, 2, 3, 4}
print("차집합(-):", a - b)   # {1}
````

> "리스트에서 중복만 빠르게 없애고 싶다" → `set()` 이 딱입니다. 단, **순서는 보장되지 않아요**.

## 6. `enumerate` 와 `zip` — 우아하게 순회하기

- **`enumerate`**: 순회하면서 **번호(인덱스)** 도 함께 준다. (출석부: 번호 + 이름)
- **`zip`**: 여러 리스트를 **나란히 묶어** 함께 순회한다. (지퍼처럼!)

````python
fruits = ["사과", "바나나", "포도"]

# enumerate : 번호와 값을 함께
for number, name in enumerate(fruits):
    print(f"{number}: {name}")

print("-" * 20)

for number, name in enumerate(fruits, start=1):   # 1번부터 시작
    print(f"{number}번째 과일은 {name}")
````

````python
# zip : 두 리스트를 나란히
names = ["지민", "하늘", "바다"]
scores = [90, 85, 95]

for name, pt in zip(names, scores):
    print(f"{name}: {pt}점")
````

## 7. Comprehension — 반복문을 한 줄로!

리스트를 만드는 `for` 문을 **한 줄**로 압축하는 문법입니다. 파이썬다움(Pythonic)의 상징이에요.

```
[식  for 변수 in 반복대상]              # 기본형
[식  for 변수 in 반복대상  if 조건]     # 조건 필터링
```

````python
# Before : for 문으로 제곱 리스트 만들기
squares = []
for n in range(1, 6):
    squares.append(n * n)
print("for 문   :", squares)

# After : 리스트 컴프리헨션 (딱 한 줄!)
squares2 = [n * n for n in range(1, 6)]
print("컴프리헨션:", squares2)
````

````python
# 조건(if) 붙이기 : 1~10 중 짝수만
evens = [n for n in range(1, 11) if n % 2 == 0]
print("짝수:", evens)
````

````python
# 문자열 리스트 가공
fruits = ["apple", "banana", "cherry"]
uppers = [f.upper() for f in fruits]
lengths = [len(f) for f in fruits]
print("대문자:", uppers)
print("길이  :", lengths)
````

````python
# 딕셔너리 컴프리헨션 : {키: 값 for ...}
square_dict = {n: n * n for n in range(1, 6)}
print("제곱사전:", square_dict)
````

## 8. lambda 와 `sorted(key)` · `map` · `filter`

- **lambda**: 이름 없는 **한 줄짜리 함수**. `lambda 매개변수: 반환식`
- **`sorted(대상, key=함수)`**: **정렬 기준**을 함수로 지정 (길이순·점수순 등)
- **`map`/`filter`**: 각 원소에 함수를 적용/조건으로 걸러내기

````python
# 일반 함수 vs lambda (같은 일)
def square(x):
    return x * x

square_lambda = lambda x: x * x
print(square(5), square_lambda(5))     # 25 25
````

````python
# sorted 의 key : 정렬 기준 정하기
words = ["banana", "kiwi", "apple", "fig"]
print("길이순  :", sorted(words, key=len))
print("알파벳순:", sorted(words))
````

````python
# 딕셔너리 리스트를 점수 기준으로 정렬 (lambda 가 진가를 발휘!)
students = [
    {"이름": "지민", "점수": 85},
    {"이름": "하늘", "점수": 92},
    {"이름": "바다", "점수": 78},
]
by_score = sorted(students, key=lambda s: s["점수"], reverse=True)
for s in by_score:
    print(s["이름"], s["점수"])
````

````python
# map : 모든 원소에 함수 적용 / filter : 조건 통과만 남기기
nums = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, nums))
evens = list(filter(lambda x: x % 2 == 0, nums))
print("map(두배) :", doubled)
print("filter(짝):", evens)
````

### 🖐️ 함께 따라하기 — 상품을 가격순으로 정렬

````python
# 🖐️ 따라하기: lambda 로 딕셔너리 리스트를 다뤄 보세요
# products = [{"이름": "펜", "가격": 1000}, {"이름": "공책", "가격": 3000}, {"이름": "지우개", "가격": 500}]
# 1) 가격이 '싼 순'(오름차순)으로 정렬해 이름과 가격 출력 (sorted, key=lambda)
# 2) filter 와 lambda 로 가격이 1000원 이상인 상품만 골라 이름 리스트 출력
````

> `map`/`filter` 로 할 수 있는 일은 대부분 **컴프리헨션**으로도 됩니다.
> `list(map(lambda x: x*2, nums))` ≡ `[x*2 for x in nums]` — 보통 컴프리헨션이 더 읽기 쉬워요.
> lambda는 특히 **`sorted`의 `key`** 로 쓸 때 가장 유용합니다.

## ⚠️ 함수와 자료구조 — 조심할 스코프 함정

교안 오전에 배웠죠: 함수 안에서 **숫자·문자열** 전역변수를 바꾸려면 `global` 선언이 필요했습니다.
그런데 **리스트·딕셔너리는 다릅니다** — `global` 없이도, 함수에 넘겨서 안에서 바꾸면 **바깥의 원본까지 바뀝니다.**

왜 그럴까요? 리스트·딕셔너리를 함수에 넘기면 **복사본이 아니라 같은 객체**가 넘어가기 때문이에요.
(이 단원 앞에서 본 `b = a` — "이름표만 하나 더" — 와 똑같은 원리입니다.)

````python
# 숫자는 안 바뀌고, 리스트는 바뀐다
def add_number(x):
    x = x + 1          # 숫자는 함수 안에서 새 값으로 갈아끼움 -> 바깥은 그대로

def add_item(items):
    items.append(99)   # 리스트는 '같은 객체'를 건드림 -> 바깥도 바뀜!

n = 10
add_number(n)
print("숫자 n   :", n)       # 10  (그대로)

nums = [1, 2, 3]
add_item(nums)
print("리스트 nums:", nums)  # [1, 2, 3, 99]  <- 넘겨준 원본이 바뀌었다!
````

> 이건 **버그이기도, 편리한 기능이기도** 합니다.
> 방금 만든 성적 관리의 `add_student(students, ...)` 가 리스트를 `return` 하지 않아도
> 명단에 학생이 쌓이는 이유가 바로 이것 — 같은 리스트를 직접 바꾸기 때문이에요.
>
> 하지만 **원본을 지켜야 할 때**는 위험합니다. 그럴 땐 함수 안에서 `.copy()` 로 복사해 쓰세요.

````python
# 원본을 지키고 싶으면: 복사본에서 작업하고 새 리스트를 return
def add_item_safe(items):
    new_items = items.copy()   # 복사본을 만들어
    new_items.append(99)       # 복사본만 바꾸고
    return new_items           # 그것을 돌려줌

data = [1, 2, 3]
result = add_item_safe(data)
print("원본 data:", data)      # [1, 2, 3]        원본은 그대로
print("결과 result:", result)  # [1, 2, 3, 99]
````
