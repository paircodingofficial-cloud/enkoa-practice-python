## 2. 함수 만들기(`def`)와 부르기(호출)

**함수를 만드는 것 = 정의(define)**,**함수를 사용하는 것 = 호출(call)** 입니다.

```python
def 함수이름():      # def 로 시작, 이름 뒤에 소괄호(), 콜론:
    실행할_코드      # 들여쓰기 한 줄 = 함수의 몸통
    return 반환할 값
```

| 용어 | 뜻 |
|---|---|
| `def` | "이제부터 함수를 정의한다"는 신호 |
| 함수이름 | 나중에 부를 때 쓰는 이름 (동사 느낌으로) |
| `()` | 값을 받을 자리 (지금은 비어 있음) |
| 들여쓰기 | 함수에 속한 코드임을 표시 (4칸) |

````python
# 정의: 배너를 찍는 함수를 '한 번만' 만들어 둡니다
def welcome_banner():
    print("=" * 24)
    print("      환영합니다!")
    print("=" * 24)

# 호출: 이름만 부르면 필요할 때마다 실행됩니다
welcome_banner()
welcome_banner()
welcome_banner()
````

> 정의만 하면 아무 일도 일어나지 않습니다. **호출(`welcome_banner()`)** 해야 비로소 실행됩니다.
> 레시피 카드를 써 두기만 하면 라면이 끓지 않듯이요!

## 3. 매개변수와 인자 — 함수에 값 건네기

지금까지 함수는 항상 똑같은 일만 했습니다. **값을 건네면** 상황에 따라 다르게 동작하게 만들 수 있어요.

>  **비유**: 자판기(함수)에 **동전과 버튼(값)** 을 넣으면 그에 맞는 음료가 나옵니다.
> 넣는 값이 달라지면 결과도 달라지죠.

| 용어 | 위치 | 뜻 |
|---|---|---|
| **매개변수**(parameter) | 함수 **정의**의 `()` 안 | 값을 받는 변수 이름 |
| **인자**(argument) | 함수 **호출**의 `()` 안 | 실제로 건네는 값 |

````python
# 이름(매개변수)을 받아 인사하는 함수
def greet(name):
    print(f"안녕하세요, {name}님!")

greet("지민")   # "지민" 이 인자 -> 매개변수 이름에 담김
greet("하늘")
greet("바다")
````

````python
# 매개변수는 여러 개도 가능 (콤마로 구분)
def introduce(name, age):
    print(f"{name}님은 {age}살이군요.")

introduce("지민", 25)
introduce("하늘", 30)
````

## 4. `return` — `print`와 무엇이 다를까? (오늘의 핵심!)

지금까지 함수는 결과를 **화면에 출력(print)** 만 했습니다. 하지만 계산 결과를 **다시 사용**하려면?

>  **비유**
> - `print` = 요리를 손님에게 **보여주기만** 함 (사진만 찍고 끝)
> - `return` = 요리를 **접시째 건네줌** → 손님이 실제로 **먹거나 다른 요리에 활용** 가능

| | `print` | `return` |
|---|---|---|
| 하는 일 | 화면에 글자를 보여줌 | 값을 함수 밖으로 **돌려줌** |
| 결과 재사용 | ❌ 불가능 | ✅ 변수에 담아 계속 사용 |
| 함수 실행 | 계속 진행 | `return`을 만나면 **즉시 종료** |

````python
# print 만 하는 함수 vs return 하는 함수 — 직접 비교!
def print_sum(a, b):
    print(a + b)      # 화면에 보여주기만

def return_sum(a, b):
    return a + b      # 값을 돌려줌

x = print_sum(3, 5)     # 화면엔 8이 찍히지만...
print("합_출력의 반환값:", x)   # x 는 None! (돌려준 게 없음)

print("-" * 30)

y = return_sum(3, 5)     # 화면엔 아무것도 안 나오지만...
print("합_반환의 반환값:", y)   # y 는 8 — 재사용 가능!
print("그 값을 두 배로:", y * 2)
print("다른 계산에도:", y + 100)
````

위 결과에서 `print_sum` 의 반환값은 `None`(아무것도 아님)이라 **재사용할 수 없습니다.**
반면 `return_sum` 은 `8`을 돌려줘서 `y * 2`, `y + 100` 처럼 **계속 활용**할 수 있죠.
**"결과를 다시 쓸 거면 `return`"** 이라고 기억하세요!

````python
# return 을 만나면 함수는 그 즉시 끝납니다 (아래 예: 절댓값)
def abs_value(n):
    if n >= 0:
        return n       # 0 이상이면 여기서 끝 (아래로 안 내려감)
    return -n          # 음수일 때만 여기 도달

print(abs_value(-7))   # 7
print(abs_value(5))    # 5
````

## 5. 기본값 매개변수 & 키워드 인자

**기본값**: 인자를 안 주면 미리 정해둔 값을 씁니다. \
**키워드 인자**: `name=값` 형태로 함수 호출 시 순서와 상관없이 함수의 인자로 건넵니다.

> ☕ **비유**: 카페 주문에서 "샷은 기본 2개" 처럼 **기본값**이 있고,
> 원하면 "샷 3개로 변경" 처럼 바꿀 수 있죠. "얼음=적게, 시럽=많이" 는 **키워드 인자**입니다.

````python
def greet(name, greeting="안녕하세요"):   # 인사말의 기본값 = "안녕하세요"
    print(f"{greeting}, {name}님!")

greet("지민")                       # 기본값 사용
greet("하늘", "반갑습니다")           # 기본값 대신 새 값
greet(name="바다", greeting="환영합니다")  # 키워드 인자 (이름을 명시)
greet(greeting="좋은 아침", name="구름")   # 키워드면 순서를 바꿔도 OK!
````

> ⚠️ **주의**: 기본값이 있는 매개변수는 함수 선언 시 **항상 뒤쪽**에 와야 합니다.
> `def f(a=1, b)` 는 에러! `def f(b, a=1)` 처럼 써야 해요.

## 6. 여러 값을 한꺼번에 반환하기

`return` 뒤에 값을 **콤마로 여러 개** 쓰면, 한 번에 여러 값을 돌려줄 수 있습니다.
받을 때도 콤마로 나눠 받으면 됩니다. (이렇게 묶인 값의 정체는 **튜플** — 이어서 공개합니다!)

````python
def quotient_remainder(a, b):
    quotient = a // b
    remainder = a % b
    return quotient, remainder        # 두 값을 한꺼번에 반환

q, r = quotient_remainder(17, 5)      # 두 변수로 나눠 받기
print(f"17 ÷ 5 = 몫 {q}, 나머지 {r}")

# min()·max() 는 여러 값 중 가장 작은/큰 값을 돌려주는 내장 함수입니다
def min_max(a, b, c):
    return min(a, b, c), max(a, b, c)

smallest, largest = min_max(7, 2, 9)
print("가장 작은 값:", smallest, "/ 가장 큰 값:", largest)
````

## 7. 변수의 범위(스코프) — 지역변수와 전역변수

함수 **안에서 만든 변수**는 그 함수 안에서만 살아있습니다(**지역변수**).
함수 **밖에서 만든 변수**는 프로그램 전체에서 보입니다(**전역변수**).

>  **비유**: 함수는 하나의 **방**입니다. 방 안에서 꺼낸 물건(지역변수)은 방 밖에서 안 보이고,
> 거실에 둔 물건(전역변수)은 어느 방에서든 볼 수 있어요.

````python
message = "전역 메시지"        # 전역변수 (함수 밖)

def show_message():
    message = "지역 메시지"    # 이 변수는 함수 '안'에서만 존재
    print("함수 안:", message)

show_message()
print("함수 밖:", message)     # 전역은 그대로 -> "전역 메시지"
````

````python
# 함수 안에서 전역변수를 '읽는' 것은 자유롭게 가능합니다
rate = 0.1

def apply_vat(price):
    return price * rate        # 함수 밖의 세율을 읽어서 사용

print(apply_vat(10000))
````

````python
# 전역변수를 함수 안에서 '수정'하려면 global 선언이 필요합니다
counter = 0

def increase():
    global counter            # "밖의 카운트를 쓰겠다" 선언
    counter += 1

increase()
increase()
increase()
print("카운트:", counter)     # 3
````

## 8. docstring & 좋은 함수 이름

- **docstring**: 함수 첫 줄에 `"""설명"""` 을 적으면 그 함수가 하는 일을 기록할 수 있습니다. `함수.__doc__` 이나 `help(함수)` 로 볼 수 있어요.
- **좋은 함수 이름**: 함수가 **하는 일**이 이름만 봐도 드러나게. 파이썬에서는 **`snake_case`**(소문자 + 밑줄)로 쓰고, 보통 **동사(+명사)** 형태로 짓습니다.

| 이 함수가 하는 일 | 나쁜 이름 | 좋은 이름 | 이유 |
|---|---|---|---|
| 가격에 부가세 10%를 더해 돌려준다 | `f`, `func1` | `add_vat` | 이름만 보고는 무슨 일인지 전혀 알 수 없음 |
| 점수 목록의 평균을 구해 돌려준다 | `data`, `avg2` | `get_average` | 명사만 있으면 "무엇을 한다"가 안 보임 — **동사로 시작** |
| 점수를 등급(A~F)으로 바꿔 돌려준다 | `process`, `doit` | `score_to_grade` | "처리한다"는 아무것도 설명하지 못함 — 무엇을 무엇으로 바꾸는지 |
| 짝수인지 판단해 True/False 를 돌려준다 | `check`, `myFunc` | `is_even` | 참·거짓을 돌려주면 `is_`/`has_` 로 시작. `myFunc` 는 camelCase — 파이썬은 snake_case |

**이름 짓기 관례 3가지**
- 동작은 **동사로 시작** — `get_`(가져오기) · `calc_`(계산) · `to_`(변환) · `add_`(더하기)
- 참·거짓을 돌려주면 **`is_` / `has_`** — `is_even`, `has_permission`
- 여러 단어는 **밑줄로** 연결 (`snake_case`) — `score_to_grade`

````python
def celsius_to_fahrenheit(c):
    """섭씨 온도 c를 화씨로 변환해 반환한다."""
    return c * 9 / 5 + 32

print(celsius_to_fahrenheit(25))
print("설명(docstring):", celsius_to_fahrenheit.__doc__)   # help(celsius_to_fahrenheit) 로도 볼 수 있어요
````

> 함수가 조금만 복잡해지면 한 줄 설명으로는 부족합니다.
> 그럴 때는 **하는 일 / `Args`(매개변수) / `Returns`(반환값)** 을 나눠 적습니다.
> 실무에서 가장 널리 쓰는 형식이라, 남이 만든 함수를 읽을 때도 이 구조로 만나게 됩니다.

````python
# 여러 줄 docstring — 하는 일 / Args(매개변수) / Returns(반환값) 을 나눠 적습니다
def calc_bmi(weight_kg, height_cm):
    """몸무게와 키로 BMI(체질량지수)를 계산해 반환한다.

    Args:
        weight_kg: 몸무게 (kg)
        height_cm: 키 (cm)

    Returns:
        BMI 값 (소수 첫째 자리까지 반올림한 실수)
    """
    height_m = height_cm / 100
    return round(weight_kg / (height_m ** 2), 1)

print("BMI:", calc_bmi(65, 170))
````

## 이번 강의 요약

| 개념 | 문법 | 핵심 |
|---|---|---|
| 함수 정의 | `def name():` | 코드를 부품으로 만들어 재사용 |
| 호출 | `name()` | 정의만으론 안 됨, 불러야 실행 |
| 매개변수·인자 | `def f(a, b)` / `f(3, 5)` | 값을 건네 다르게 동작 |
| **return** | `return 값` | **결과를 돌려줘 재사용** (print와 구분!) |
| 기본값 | `def f(a=1)` | 인자 생략 시 기본값 |
| 키워드 인자 | `f(a=1, b=2)` | 순서 무관하게 명시 |
| 다중 반환 | `return a, b` | 여러 값을 한꺼번에 (튜플) |
| 스코프 | 지역/전역, `global` | 함수 안 변수는 밖에서 안 보임 |
| docstring | `"""설명"""` | 함수 설명 기록 |
