**예시**

## 8. 시저 암호 (암호화 · 복호화)

**배경**: 각 알파벳을 정해진 칸수만큼 밀어 암호화하는 고전 암호입니다(율리우스 카이사르가 썼다고 전해집니다).

**요구사항**: 영문 **소문자**로 된 문장을 받아
1. `caesar_encrypt(sentence, shift=3)` : 각 소문자를 `shift` 칸 뒤로 민다(z 다음은 a 로 순환). 소문자가 아닌 글자(공백 등)는 그대로 둔다.
2. `caesar_decrypt(sentence, shift=3)` : 암호화를 되돌린다.

**예시**
```
caesar_encrypt("abc")         →  "def"
caesar_encrypt("xyz")         →  "abc"   (순환)
caesar_encrypt("hello", 5)    →  "mjqqt"
caesar_decrypt("mjqqt", 5)    →  "hello"
```

<details><summary>힌트</summary>

```text
접근방법:
- 각 글자를 알파벳 순번(0~25)으로 바꿔 shift 만큼 민 뒤, 26으로 나눈 나머지로 순환시키고 다시 글자로 되돌림

세부구현:
1. caesar_encrypt(sentence, shift=3)
   1-1. 빈 결과 문자열을 준비
   1-2. 문장의 각 글자에 대해, 소문자(a~z)이면 아래 순서로 바꿔 결과에 붙임
        (1) 그 글자가 'a'로부터 몇 번째인지 순번을 구함 (a=0, b=1 …, ord 로)
        (2) 순번에 shift 를 더한 뒤 26으로 나눈 나머지를 취함 (z를 넘으면 a로 돌아옴)
        (3) 그 순번을 다시 글자로 되돌림 (chr 로, 'a'의 코드에 순번을 더해)
   1-3. 소문자가 아니면(공백 등) 원래 글자를 그대로 붙임
   1-4. 결과 문자열을 반환
2. caesar_decrypt(sentence, shift=3)
   2-1. 미는 방향만 반대 — shift 를 음수로 주어 caesar_encrypt 를 그대로 재사용

참고:
- 글자와 순번 변환에는 ord / chr 을 사용
- 파이썬의 나머지 연산은 음수에도 안전해 복호화도 같은 공식으로 동작
```

</details>

````python
# 여기에 코드를 작성하세요
````

````python
# [자가채점]
assert caesar_encrypt("abc") == "def"
assert caesar_encrypt("xyz") == "abc"          # z 다음은 a 로 순환
assert caesar_encrypt("hello", 5) == "mjqqt"
assert caesar_decrypt(caesar_encrypt("python")) == "python"
assert caesar_encrypt("ai camp", 1) == "bj dbnq"   # 공백은 유지
print("통과! ")
````

## 9. 상위 3명 뽑기

**배경**: 점수 딕셔너리에서 **점수가 높은 상위 3명**의 이름을 뽑습니다. `sorted` 와 `lambda` 를 함께 활용합니다.

**요구사항**: `{name: score}` 딕셔너리를 받아 **점수 내림차순 상위 3명의 이름 리스트**를 반환하는 `top3(score_dict)` 함수를 작성하세요.
- 인원이 3명보다 적으면 **있는 만큼만** 반환합니다(2명이면 이름 2개).

```
score = {"지민": 85, "하늘": 92, "바다": 78, "구름": 95, "별": 60}
top3(score)  →  ["구름", "하늘", "지민"]
```

```text
세부구현:
1. 딕셔너리의 항목들(items)을 꺼냄
2. 정렬 기준(key)을 "각 쌍의 점수"로, 방향을 내림차순(reverse)으로 지정해 정렬
```

## 10. 취미 비교하기 (set 집합 연산 — 교집합·차집합·합집합)

**배경**: 5번에서는 `set` 의 **교집합(`&`)** 으로 공통 원소를 구했죠. 이번엔 여기에 **차집합(`-`)** 과 **합집합(`|`)** 까지 더해, 두 사람의 취미를 세 가지 각도로 비교합니다.

**요구사항**: 두 리스트 `mine`, `yours` 를 받아 `(공통, 나만, 전체)` **튜플**을 반환하는 `hobby_compare(mine, yours)` 함수를 작성하세요.
- 공통: 두 사람 **모두** 가진 취미 — 교집합 (`&`)
- 나만: `mine` 에만 있고 `yours` 엔 없는 취미 — 차집합 (`-`)
- 전체: 두 사람의 취미를 **다 합친** 목록(중복 없이) — 합집합 (`|`)
- `set` 은 순서가 없으므로 **세 결과 모두 `sorted` 로 정렬**해 리스트로 반환합니다.

**예시**
```
hobby_compare(["독서", "게임", "등산"], ["게임", "요리", "등산"])
    →  (["게임", "등산"], ["독서"], ["게임", "독서", "등산", "요리"])
```
