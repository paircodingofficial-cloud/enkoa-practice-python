## 1. 연락처 관리 프로그램

**배경**: 이름으로 전화번호를 저장·검색·삭제하는 미니 연락처 앱을 만듭니다.
저장소는 `{name: 전화번호}` 딕셔너리, 각 기능은 **함수**로, 사용자 조작은 **while 메뉴 루프**로 처리합니다.

**단계별 목표**
- **v1**: `add_contact`, `전체보기` 기능부터
- **v2**: `find_contact`, `delete_contact` 추가
- **v3**: 예외 상황 방어 — 이미 있는 이름을 추가하면 "수정" 처리, 없는 이름을 검색·삭제하면 안내

**만들 함수** (모두 첫 인자로 연락처 딕셔너리를 받습니다)

| 함수 | 하는 일 | 반환 |
|---|---|---|
| `add_contact(contacts, name, number)` | 저장(있으면 수정) | 안내 메시지 문자열 |
| `find_contact(contacts, name)` | 번호 찾기 | 번호(없으면 `None`) |
| `delete_contact(contacts, name)` | 삭제 | 성공 `True` / 없으면 `False` |

> `add_contact` 반환 문자열은 정확히 — 새 이름이면 `"{name}님을 추가했습니다."`, 이미 있는 이름이면 `"{name}님의 번호를 수정했습니다."` (예: 지민 추가 시 `"지민님을 추가했습니다."`)

**메뉴 루프**: `1.추가  2.검색  3.삭제  4.전체보기  0.종료` (0을 누르면 종료)

<details><summary>힌트</summary>

```text
접근방법:
- 데이터(딕셔너리)와 기능(함수)을 분리
- 세 함수는 모두 연락처 딕셔너리를 인자로 받아 동작하고, 사용자 입력은 while 메뉴 루프가 담당

세부구현:
1. add_contact(contacts, name, number)
   1-1. 이름이 이미 있으면 "수정" 안내 메시지, 없으면 "추가" 안내 메시지를 준비
   1-2. 어느 경우든 그 이름 키에 번호를 저장하고, 준비한 메시지를 반환
2. find_contact(contacts, name)
   2-1. 이름이 있으면 그 번호를, 없으면 None 을 반환
        - 없을 때 None 을 돌려주는 조회 메서드(get)를 쓰면 간단
3. delete_contact(contacts, name)
   3-1. 이름이 있으면 그 항목을 지우고 True, 없으면 False 를 반환
4. 메뉴 루프
   4-1. 무한 반복을 열고, 매 회 메뉴를 출력한 뒤 번호를 입력받음
   4-2. 0이면 반복을 끝냄 (종료)
   4-3. 1~3이면 이름(과 번호)을 추가로 입력받아 해당 함수를 호출하고 결과를 안내
   4-4. 4면 딕셔너리 전체를 순회하며 이름·번호를 출력
   4-5. 그 외 번호는 잘못된 입력으로 안내
```
