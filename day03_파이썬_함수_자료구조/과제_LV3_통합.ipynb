## 1. 연락처 관리 프로그램

**배경**: 이름으로 전화번호를 저장·검색·삭제하는 미니 연락처 앱을 만듭니다.
저장소는 `{name: 전화번호}` 딕셔너리, 각 기능은 **함수**로, 사용자 조작은 **while 메뉴 루프**로 처리합니다.

**단계별 목표**
- **v1**: `add_contact`, `전체보기` 기능부터
- **v2**: `find_contact`, `delete_contact` 추가
- **v3**: 예외 상황 방어 — 이미 있는 이름을 추가하면 "수정" 처리, 없는 이름을 검색·삭제하면 안내

**만들 함수** (모두 첫 인자로 연락처 딕셔너리를 받습니다)

| 함수 | 하는 일 | 반환 |
|---|---|---|
| `add_contact(contacts, name, number)` | 저장(있으면 수정) | 안내 메시지 문자열 |
| `find_contact(contacts, name)` | 번호 찾기 | 번호(없으면 `None`) |
| `delete_contact(contacts, name)` | 삭제 | 성공 `True` / 없으면 `False` |

> `add_contact` 반환 문자열은 정확히 — 새 이름이면 `"{name}님을 추가했습니다."`, 이미 있는 이름이면 `"{name}님의 번호를 수정했습니다."` (예: 지민 추가 시 `"지민님을 추가했습니다."`)

**메뉴 루프**: `1.추가  2.검색  3.삭제  4.전체보기  0.종료` (0을 누르면 종료)

<details><summary>힌트</summary>

```text
접근방법:
- 데이터(딕셔너리)와 기능(함수)을 분리
- 세 함수는 모두 연락처 딕셔너리를 인자로 받아 동작하고, 사용자 입력은 while 메뉴 루프가 담당

세부구현:
1. add_contact(contacts, name, number)
   1-1. 이름이 이미 있으면 "수정" 안내 메시지, 없으면 "추가" 안내 메시지를 준비
   1-2. 어느 경우든 그 이름 키에 번호를 저장하고, 준비한 메시지를 반환
2. find_contact(contacts, name)
   2-1. 이름이 있으면 그 번호를, 없으면 None 을 반환
        - 없을 때 None 을 돌려주는 조회 메서드(get)를 쓰면 간단
3. delete_contact(contacts, name)
   3-1. 이름이 있으면 그 항목을 지우고 True, 없으면 False 를 반환
4. 메뉴 루프
   4-1. 무한 반복을 열고, 매 회 메뉴를 출력한 뒤 번호를 입력받음
   4-2. 0이면 반복을 끝냄 (종료)
   4-3. 1~3이면 이름(과 번호)을 추가로 입력받아 해당 함수를 호출하고 결과를 안내
   4-4. 4면 딕셔너리 전체를 순회하며 이름·번호를 출력
   4-5. 그 외 번호는 잘못된 입력으로 안내
```

</details>

````python
# [자가채점]
phonebook = {}
assert add_contact(phonebook, "지민", "010-1111") == "지민님을 추가했습니다."
assert phonebook["지민"] == "010-1111"
# 이미 있는 이름 -> 수정 처리 (v3 예외 방어)
assert add_contact(phonebook, "지민", "010-2222") == "지민님의 번호를 수정했습니다."
assert phonebook["지민"] == "010-2222"
# 검색
assert find_contact(phonebook, "지민") == "010-2222"
assert find_contact(phonebook, "없는사람") is None       # 없는 이름 검색
# 삭제
assert delete_contact(phonebook, "지민") is True
assert delete_contact(phonebook, "지민") is False          # 이미 삭제됨
assert len(phonebook) == 0
print("통과! ")
````

## 2. 가위바위보 전적 관리

**배경**: 컴퓨터와 5판 가위바위보를 하고 **전적(승/무/패)과 승률**을 집계합니다.

**규칙**: `1=가위, 2=바위, 3=보`. 이기는 관계는 가위>보, 바위>가위, 보>바위.

**만들 것**
1. `judge(player, computer)` : 두 선택(1/2/3)을 받아 `"승"`, `"무"`, `"패"` 중 하나 반환
2. 전적 딕셔너리 `{"승": 0, "무": 0, "패": 0}` 를 두고, **5판**을 돌며 결과를 누적
3. 5판 후 `3승 1무 1패` 형식으로 전적을, 이어서 `승률: 60%` 형식으로 승률을 출력하세요 (**소수점 없이 반올림** — `:.0f`)

> 컴퓨터의 선택은 아래 **[제공 코드]** 의 `computer_choice()` 를 그대로 사용하세요.
> 부를 때마다 **무작위로** 가위·바위·보(`1`/`2`/`3`) 중 하나를 냅니다 — 그래서 매번 결과가 달라져요!

## 3. 학생 성적 관리 프로그램

**배경**: 학생 한 명을 `{"name": 이름, "score": 점수}` **딕셔너리**로 만들어 **리스트에 담아** 관리합니다.
교안에서 "실무 데이터는 대부분 이 모양"이라고 했던 **리스트 안의 딕셔너리** 구조예요.

**만들 함수 4개**

| 함수 | 하는 일 | 반환 |
|---|---|---|
| `add_student(students, name, score)` | 명단 끝에 학생 추가 | `"지민 학생을 추가했습니다."` 형식 문자열 |
| `find_score(students, name)` | 이름으로 점수 찾기 | 점수(정수). 없으면 `None` |
| `class_average(students)` | 평균 점수 | 실수, **소수 첫째 자리까지 반올림**. **빈 명단이면 `0.0`** |
| `ranking(students)` | 점수 내림차순 순위 | `["1위 하늘(92점)", "2위 지민(85점)"]` 형식 리스트 |

**메뉴 루프**: `1.추가  2.검색  3.평균  4.순위  0.종료` — **메뉴는 번호**로, **학생 이름은 글자로** 입력받으세요(1번 연락처 문제와 같은 방식).

**예시**
```
students = [{"name": "지민", "score": 85}, {"name": "하늘", "score": 92}]

class_average(students)   ->  88.5
find_score(students, "하늘")  ->  92
find_score(students, "없음")  ->  None
ranking(students)         ->  ["1위 하늘(92점)", "2위 지민(85점)"]
```

<details><summary>힌트</summary>

```text
접근방법:
- 학생 하나 = 딕셔너리, 명단 = 그 딕셔너리들을 담은 리스트.
- 리스트를 for 로 훑으며 각 딕셔너리의 "name"·"score" 를 꺼내 쓴다.

세부구현:
1. add_student
   1-1. {"name": name, "score": score} 를 만들어 리스트에 append
   1-2. f-string 으로 안내 문구를 만들어 반환
2. find_score
   2-1. 명단을 훑으며 이름이 같은 학생을 만나면 그 점수를 반환
   2-2. 끝까지 못 찾으면 None 반환
3. class_average
   3-1. 명단이 비었으면 0.0 을 먼저 반환 (0으로 나누는 에러 방지)
   3-2. 점수만 모아 합을 구하고 인원으로 나눈 뒤 round(값, 1)
4. ranking
   4-1. sorted 에 key=lambda 로 "점수"를 기준, reverse=True 로 내림차순
   4-2. enumerate(정렬결과, start=1) 로 번호를 붙여 f-string 문자열 리스트로
```

</details>

````python
# [자가채점]
students = []
assert add_student(students, "지민", 85) == "지민 학생을 추가했습니다."
add_student(students, "하늘", 92)
assert len(students) == 2
assert students[0] == {"name": "지민", "score": 85}

assert find_score(students, "하늘") == 92
assert find_score(students, "없음") is None

assert class_average(students) == 88.5
assert class_average([]) == 0.0

assert ranking(students) == ["1위 하늘(92점)", "2위 지민(85점)"]
print("통과! ")
````
