### 이미지 검색
[이미지 API 문서 링크](https://developers.naver.com/docs/serviceapi/search/image/image.md#%EC%9D%B4%EB%AF%B8%EC%A7%80)

(1) 기본 요청 받기

(2) 이미지 다운로드

```python
import os
import re
import requests

def get_filename(filename) :
    """파일명에 사용할 수 없는 특수문자 제거"""
    return re.sub(r'[\\/:*?"<>|]', '', filename)

def download_image(url, folder_path, filename):
    """이미지를 다운로드하는 함수"""
    try:
        with requests.get(url, stream=True, timeout=10) as response:
            if response.status_code != 200:
                return False, f"HTTP Error: {response.status_code}"

            content_type = response.headers.get('content-type', '')
            ext = content_type.split('/')[-1] if 'image' in content_type else 'jpg'

            final_path = os.path.join(folder_path, f"{filename}.{ext}")
            with open(final_path, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)

            return True, final_path
    except Exception as e:
        return False, str(e)
```
